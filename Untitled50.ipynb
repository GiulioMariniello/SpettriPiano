```python
# -*- coding: utf-8 -*-

import os
import re
import shutil
from io import StringIO

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.signal import welch
from scipy.integrate import cumulative_trapezoid

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e


# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81  # m/s^2

FLOORS_TO_PLOT = [3, 5, 10, 15]

# 1 file per N_piani (personalizza coi tuoi nomi reali)
EXCEL_BY_FLOORS = {
    3:  "RISULTATI_3P.xlsx",
    5:  "RISULTATI_5P.xlsx",
    10: "RISULTATI_10P.xlsx",
    15: "RISULTATI_15P.xlsx",
}

# finestra di analisi comune per TUTTO (TH/PSD/SA/Husid)
TMAX = 30.0

# Input base: COLB preferito
INPUT_COLB = "COLB.txt"
INPUT_DT = 0.01  # dt del COLB (se diverso, cambialo)

# Spettri
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05
SA_XMAX = 3.0

# PSD
PSD_FMAX = 50.0
WELCH_NPERSEG = 2048  # adeguabile

# Output
OUTDIR = "POSTPROCESS_OUTPUT"
DIR_COMPARE = os.path.join(OUTDIR, "01_CONFRONTI_TH_PSD_SA")
DIR_HUSID = os.path.join(OUTDIR, "02_HUSID_ARIAS")
DIR_TABLES = os.path.join(OUTDIR, "03_TABELLE")
DIR_REDUCTIONS = os.path.join(OUTDIR, "04_RIDUZIONI_ISTOGRAMMI")


# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df


def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


def extract_tis(model_name: str) -> float:
    m = re.search(r"(\d+\.\d+)", str(model_name))
    return float(m.group(1)) if m else 0.0


def classify_strategy(tipo_base: str, modello: str):
    tb = (tipo_base or "").upper()
    mn = (modello or "").upper()
    txt = tb if tb not in ("", "NAN", "NONE") else mn

    if ("RIGIDA" in txt) or ("FISSA" in txt) or ("FIXED" in txt):
        return "RIGIDA"
    if ("SLITT" in txt) or ("FRICTION" in txt):
        return "SLITTE"
    if ("ISOL" in txt) or ("ELAST" in txt):
        return "ISOLATA"
    return "ALTRO"


def style_for_category(cat: str, tis: float):
    if cat == "RIGIDA":
        return dict(color="black", lw=2.8, z=10, label="Base Fissa")
    if cat == "SLITTE":
        return dict(color="#D55E00", lw=1.8, z=5, label=f"Isolato+Slitte T={tis:.1f}s")
    if cat == "ISOLATA":
        return dict(color="#0072B2", lw=1.8, z=5, label=f"Isolato T={tis:.1f}s")
    return dict(color="gray", lw=1.2, z=1, label="Altro")


def sort_models_for_plot(df_models: pd.DataFrame) -> pd.DataFrame:
    order_map = {"RIGIDA": 0, "SLITTE": 1, "ISOLATA": 2, "ALTRO": 9}
    tmp = df_models.copy()
    tmp["__ord"] = tmp["Categoria"].map(order_map).fillna(9).astype(int)
    return tmp.sort_values(["__ord", "Tis", "Modello"]).drop(columns="__ord")


def truncate_by_time(t: np.ndarray, x: np.ndarray, tmax: float):
    if len(t) == 0:
        return t, x
    mask = t <= tmax
    return t[mask], x[mask]


# ============================================================
# 2) INPUT BASE: leggo solo PGA per asse (niente TH input)
# ============================================================
def read_colb_pga(path: str, dt_input: float):
    if not os.path.exists(path):
        return {}

    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    if df.shape[1] < 3:
        return {}

    acc = {
        "HNE": df.iloc[:, 0].to_numpy(dtype=float) * G,
        "HNN": df.iloc[:, 1].to_numpy(dtype=float) * G,
        "HNZ": df.iloc[:, 2].to_numpy(dtype=float) * G,
    }

    pga = {}
    for ax, a in acc.items():
        t = np.arange(len(a)) * dt_input
        t, a = truncate_by_time(t, a, TMAX)
        pga[ax] = float(np.max(np.abs(a))) if len(a) else 0.0
    return pga


# ============================================================
# 3) ARIAS + HUSID (solo 0-30 s)
# ============================================================
def arias_curve(acc: np.ndarray, t: np.ndarray) -> np.ndarray:
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ


def husid_metrics(Ia_curve: np.ndarray, t: np.ndarray, p1=0.05, p2=0.95):
    Ia_tot = float(Ia_curve[-1]) if len(Ia_curve) else 0.0
    if Ia_tot <= 0:
        return dict(Ia_total=Ia_tot, t1=np.nan, t2=np.nan, duration=np.nan, husid=None)

    h = Ia_curve / Ia_tot

    def _time_at(p):
        idx = np.searchsorted(h, p, side="left")
        if idx <= 0:
            return float(t[0])
        if idx >= len(t):
            return float(t[-1])
        t0, t1 = float(t[idx - 1]), float(t[idx])
        h0, h1 = float(h[idx - 1]), float(h[idx])
        if h1 == h0:
            return t1
        return t0 + (p - h0) * (t1 - t0) / (h1 - h0)

    tp1 = _time_at(p1)
    tp2 = _time_at(p2)
    return dict(Ia_total=Ia_tot, t1=tp1, t2=tp2, duration=tp2 - tp1, husid=h)


# ============================================================
# 4) LETTURA EXCEL
# ============================================================
def load_excel_bundle(excel_files):
    list_r, list_th = [], []

    for f in excel_files:
        if not os.path.exists(f):
            print(f"⚠️ File non trovato: {f}")
            continue

        xls = pd.ExcelFile(f)
        if "Riepilogo" in xls.sheet_names:
            list_r.append(pd.read_excel(xls, sheet_name="Riepilogo"))
        if "TimeHistory" in xls.sheet_names:
            list_th.append(pd.read_excel(xls, sheet_name="TimeHistory"))

    if not list_r:
        raise ValueError("❌ Nessun foglio 'Riepilogo' caricato dagli Excel.")

    df_r = pd.concat(list_r, ignore_index=True)
    df_th = pd.concat(list_th, ignore_index=True) if list_th else pd.DataFrame()

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Tipo_Base", "Nodo_Tag", "Nodo_ID"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])

    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    if set(["Modello", "Nodo_ID", "Asse"]).issubset(df_r.columns):
        df_r = df_r.drop_duplicates(subset=["Modello", "Nodo_ID", "Asse"])

    if "Tipo_Base" not in df_r.columns:
        df_r["Tipo_Base"] = ""

    df_r["Tis"] = df_r["Modello"].apply(extract_tis)
    df_r["Categoria"] = df_r.apply(lambda r: classify_strategy(r.get("Tipo_Base", ""), r.get("Modello", "")), axis=1)

    return df_r, df_th


# ============================================================
# 5) PSD/SA coerenti su 0-30 s (calcolate on-the-fly)
# ============================================================
def compute_psd_welch(t: np.ndarray, a: np.ndarray):
    if len(t) < 4:
        return np.array([]), np.array([])
    dt = float(np.mean(np.diff(t)))
    fs = 1.0 / dt if dt > 0 else 0.0
    if fs <= 0:
        return np.array([]), np.array([])
    f, Pxx = welch(a, fs=fs, nperseg=min(WELCH_NPERSEG, len(a)))
    mask = f <= PSD_FMAX
    return f[mask], Pxx[mask]


def compute_sa(t: np.ndarray, a: np.ndarray, cache: dict, key):
    if key in cache:
        return cache[key]
    if len(t) < 2:
        sa = np.zeros_like(PERIODS)
        cache[key] = sa
        return sa
    dt = float(np.mean(np.diff(t)))
    spec = pyrotd.calc_spec_accels(dt, a, OSC_FREQS, OSC_DAMPING)
    sa = np.asarray(spec.spec_accel, dtype=float)
    cache[key] = sa
    return sa


# ============================================================
# 6) PREPASS: limiti globali uguali per ogni tipo di plot
# ============================================================
def compute_global_limits(df_r, df_th, pga_input_by_axis):
    """
    Calcola massimi globali (considerando solo 0-30 s) per:
    - TH (acc)
    - PSD
    - SA (solo HNE/HNN)
    """
    max_acc = 0.0
    max_psd = 0.0
    max_sa = 0.0

    # includo anche PGA input nelle scale TH
    for ax, p in pga_input_by_axis.items():
        max_acc = max(max_acc, float(p))

    sa_cache = {}

    groups = df_r[["N_Piani", "Nodo_ID", "Asse", "Modello"]].drop_duplicates()

    for _, g in groups.iterrows():
        mod = g["Modello"]
        node_id = g["Nodo_ID"]
        axis = g["Asse"]

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        if len(a):
            max_acc = max(max_acc, float(np.max(np.abs(a))))

        f, Pxx = compute_psd_welch(t, a)
        if len(Pxx):
            max_psd = max(max_psd, float(np.max(Pxx)))

        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            if len(sa):
                max_sa = max(max_sa, float(np.max(sa)))

    # margini
    lims = {
        "th_y": 1.15 * max_acc if max_acc > 0 else 1.0,
        "psd_y": 1.15 * max_psd if max_psd > 0 else 1.0,
        "sa_y": 1.15 * max_sa if max_sa > 0 else 1.0,
    }
    return lims


# ============================================================
# 7) PLOT: TRIPTICO + HUSID (con limiti globali)
# ============================================================
def plot_triptych(nP, node_id, node_tag, axis, df_models, df_th, pga_input_by_axis, sa_cache, lims):
    fig, axs = plt.subplots(1, 3, figsize=(24, 7))
    fig.suptitle(f"{nP} Piani | {node_tag} ({node_id}) | {axis}", fontsize=16, fontweight="bold")

    # --- TH: solo linee PGA input (niente TH input) ---
    pga_in = float(pga_input_by_axis.get(axis, 0.0))
    if pga_in > 0:
        axs[0].axhline(+pga_in, color="gray", ls="--", lw=1.6, alpha=0.85, label="±PGA Input")
        axs[0].axhline(-pga_in, color="gray", ls="--", lw=1.6, alpha=0.85)

    df_models = sort_models_for_plot(df_models)

    for _, r in df_models.iterrows():
        mod = r["Modello"]
        tis = float(r.get("Tis", 0.0))
        cat = r.get("Categoria", "ALTRO")
        st = style_for_category(cat, tis)

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        axs[0].plot(t, a, color=st["color"], lw=st["lw"], alpha=0.88, zorder=st["z"], label=st["label"])

        # PSD calcolata su 0-30 s
        f, Pxx = compute_psd_welch(t, a)
        if len(f):
            axs[1].plot(f, Pxx, color=st["color"], lw=1.6, alpha=0.9, zorder=st["z"])

        # SA (solo HNE/HNN)
        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            axs[2].plot(PERIODS, sa, color=st["color"], lw=st["lw"], alpha=0.9, zorder=st["z"])

    # formatting (assi globali)
    axs[0].set_title("Time History (0–30 s) + ±PGA input", fontsize=13)
    axs[0].set_xlabel("Tempo [s]")
    axs[0].set_ylabel("Acc [m/s²]")
    axs[0].set_xlim(0, TMAX)
    axs[0].set_ylim(-lims["th_y"], lims["th_y"])
    axs[0].grid(True, ls=":", alpha=0.5)
    axs[0].legend(loc="upper right", fontsize=10)

    axs[1].set_title("PSD (Welch) su 0–30 s", fontsize=13)
    axs[1].set_xlabel("Frequenza [Hz]")
    axs[1].set_xlim(0, PSD_FMAX)
    axs[1].set_ylim(0, lims["psd_y"])
    axs[1].grid(True, ls=":", alpha=0.5)

    if axis != "HNZ":
        axs[2].set_title("Spettro (5%) su 0–30 s", fontsize=13)
        axs[2].set_xlabel("Periodo [s]")
        axs[2].set_ylabel("SA [m/s²]")
        axs[2].set_xlim(0, SA_XMAX)
        axs[2].set_ylim(0, lims["sa_y"])
        axs[2].grid(True, ls=":", alpha=0.5)
    else:
        axs[2].set_visible(False)

    return fig


def plot_husid(nP, node_id, node_tag, axis, df_models, df_th, pga_input_by_axis):
    fig, ax = plt.subplots(figsize=(12, 7))
    fig.suptitle(f"Husid (Arias normalizzata) | {nP} Piani | {node_tag} ({node_id}) | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    rows = []
    df_models = sort_models_for_plot(df_models)

    for _, r in df_models.iterrows():
        mod = r["Modello"]
        tis = float(r.get("Tis", 0.0))
        cat = r.get("Categoria", "ALTRO")
        st = style_for_category(cat, tis)

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        Ia = arias_curve(a, t)
        m = husid_metrics(Ia, t, 0.05, 0.95)
        if m["husid"] is None:
            continue

        ax.plot(t, m["husid"], color=st["color"], lw=st["lw"], alpha=0.9,
                label=f"{st['label']} | D5-95={m['duration']:.2f}s")

        rows.append({
            "N_Piani": nP,
            "Nodo_ID": node_id,
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "Modello": mod,
            "Categoria": cat,
            "Tis": tis,
            "Arias_Total": m["Ia_total"],
            "t5_s": m["t1"],
            "t95_s": m["t2"],
            "Durata_5_95_s": m["duration"],
            "PGA_Input": float(pga_input_by_axis.get(axis, 0.0)),
        })

    ax.set_xlim(0, TMAX)
    ax.set_ylim(0, 1.02)
    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Husid = Ia(t)/Ia(T)")
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="lower right", fontsize=9, framealpha=0.95)

    return fig, rows


# ============================================================
# 8) RIDUZIONI + ISTOGRAMMI
# ============================================================
def build_results_table(df_r, df_th):
    """
    Costruisce una tabella per ogni (N_Piani, Nodo_ID, Asse, Modello):
    - PGA_Max (0-30 s)
    - SA_Max (0-30 s, solo HNE/HNN)
    """
    sa_cache = {}
    results = []

    groups = df_r[["N_Piani", "Nodo_ID", "Nodo_Tag", "Asse", "Modello", "Categoria", "Tis"]].drop_duplicates()

    for _, g in groups.iterrows():
        nP = int(g["N_Piani"])
        node_id = g["Nodo_ID"]
        node_tag = g["Nodo_Tag"]
        axis = g["Asse"]
        mod = g["Modello"]
        cat = g["Categoria"]
        tis = float(g["Tis"])

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        pga = float(np.max(np.abs(a))) if len(a) else 0.0

        sa_max = 0.0
        if axis != "HNZ" and len(a) >= 2:
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            sa_max = float(np.max(sa)) if len(sa) else 0.0

        results.append({
            "N_Piani": nP,
            "Nodo_ID": node_id,
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "Modello": mod,
            "Categoria": cat,
            "Tis": tis,
            "PGA_Max": pga,
            "SA_Max": sa_max,
        })

    return pd.DataFrame(results)


def compute_reductions(df_res, pga_input_by_axis):
    """
    Riduzioni % rispetto a RIGIDA per ogni (N_Piani, Nodo_ID, Asse)
    """
    final_rows = []

    grouped = df_res.groupby(["N_Piani", "Nodo_ID", "Asse"])
    for (nP, nID, ax), group in grouped:
        ref = group[group["Categoria"] == "RIGIDA"]
        if ref.empty:
            continue

        ref_pga = float(ref.iloc[0]["PGA_Max"])
        ref_sa = float(ref.iloc[0]["SA_Max"])

        for _, row in group.iterrows():
            if row["Categoria"] == "RIGIDA":
                continue

            red_pga = (1 - float(row["PGA_Max"]) / ref_pga) * 100 if ref_pga > 0 else 0.0
            red_sa  = (1 - float(row["SA_Max"])  / ref_sa)  * 100 if ref_sa > 0 else 0.0

            tis_label = f"{row['Tis']:.1f}s" if pd.notna(row["Tis"]) else "N/A"
            strategy_label = ("Slitte " if row["Categoria"] == "SLITTE" else "Isolato ") + tis_label

            final_rows.append({
                "N_Piani": int(nP),
                "Nodo_ID": nID,
                "Asse": ax,
                "Categoria": row["Categoria"],
                "Strategia": strategy_label,
                "Tis": float(row["Tis"]),
                "Modello": row["Modello"],
                "PGA_Input": float(pga_input_by_axis.get(ax, 0.0)),
                "PGA_Output": float(row["PGA_Max"]),
                "Riduzione_PGA_%": round(float(red_pga), 2),
                "SA_Output_Max": float(row["SA_Max"]),
                "Riduzione_SA_%": round(float(red_sa), 2),
            })

    return pd.DataFrame(final_rows)


def plot_reduction_histograms(df_red):
    """
    Istogrammi (bar charts) delle riduzioni medie per piano:
    - riduzione PGA
    - riduzione SA (assi HNE/HNN)
    Barre raggruppate per asse; categorie/strategie ordinate per Tis.
    """
    if df_red.empty:
        return

    os.makedirs(DIR_REDUCTIONS, exist_ok=True)

    df_plot = df_red.copy()

    # Ordina strategie: prima SLITTE poi ISOLATA e per Tis crescente (coerente col tuo codice)
    df_plot["SortKey"] = df_plot["Tis"]
    df_plot = df_plot.sort_values(by=["Categoria", "SortKey"])

    for nP in sorted(df_plot["N_Piani"].unique()):
        sub = df_plot[df_plot["N_Piani"] == nP].copy()
        if sub.empty:
            continue

        # media su tutti i nodi (come nel tuo script)
        summary = sub.groupby(["Asse", "Strategia"])[["Riduzione_PGA_%", "Riduzione_SA_%"]].mean().reset_index()

        strategies = sorted(summary["Strategia"].unique(), key=lambda s: (1 if "Isolato" in s else 0, extract_tis(s)))
        assi = ["HNE", "HNN", "HNZ"]

        fig, axs = plt.subplots(1, 2, figsize=(18, 9))
        fig.suptitle(f"Efficienza Isolamento (0–30 s) - {nP} Piani", fontsize=18, fontweight="bold")

        width = 0.25
        x = np.arange(len(strategies))

        # --- Riduzione PGA ---
        for i, ax_code in enumerate(assi):
            vals = []
            for s in strategies:
                row = summary[(summary["Asse"] == ax_code) & (summary["Strategia"] == s)]
                vals.append(float(row["Riduzione_PGA_%"].values[0]) if not row.empty else 0.0)

            offset = (i - 1) * width
            rects = axs[0].bar(x + offset, vals, width, label=ax_code, alpha=0.85, edgecolor="black")
            axs[0].bar_label(rects, fmt="%.0f%%", padding=3, fontsize=9, rotation=90)

        axs[0].set_title("Riduzione PGA Output vs Base Fissa", fontsize=13)
        axs[0].set_ylabel("Riduzione [%]", fontsize=12)
        axs[0].set_xticks(x)
        axs[0].set_xticklabels(strategies, rotation=45, ha="right", fontsize=10)
        axs[0].set_ylim(0, 100)
        axs[0].grid(axis="y", ls="--", alpha=0.5)
        axs[0].legend(loc="lower left")

        # --- Riduzione SA (solo HNE/HNN) ---
        assi_sa = ["HNE", "HNN"]
        for i, ax_code in enumerate(assi_sa):
            vals = []
            for s in strategies:
                row = summary[(summary["Asse"] == ax_code) & (summary["Strategia"] == s)]
                vals.append(float(row["Riduzione_SA_%"].values[0]) if not row.empty else 0.0)

            offset = (i - 0.5) * width
            rects = axs[1].bar(x + offset, vals, width, label=ax_code, alpha=0.85, edgecolor="black")
            axs[1].bar_label(rects, fmt="%.0f%%", padding=3, fontsize=9, rotation=90)

        axs[1].set_title("Riduzione SA Max Output vs Base Fissa (HNE/HNN)", fontsize=13)
        axs[1].set_ylabel("Riduzione [%]", fontsize=12)
        axs[1].set_xticks(x)
        axs[1].set_xticklabels(strategies, rotation=45, ha="right", fontsize=10)
        axs[1].set_ylim(0, 100)
        axs[1].grid(axis="y", ls="--", alpha=0.5)
        axs[1].legend(loc="lower left")

        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_REDUCTIONS, f"ISTOGRAMMI_RIDUZIONI_{nP}P.png"), dpi=150)
        plt.close(fig)


# ============================================================
# 9) MAIN
# ============================================================
def main():
    # reset output
    if os.path.exists(OUTDIR):
        shutil.rmtree(OUTDIR)
    os.makedirs(DIR_COMPARE, exist_ok=True)
    os.makedirs(DIR_HUSID, exist_ok=True)
    os.makedirs(DIR_TABLES, exist_ok=True)
    os.makedirs(DIR_REDUCTIONS, exist_ok=True)

    # PGA input (0–30 s)
    pga_input_by_axis = read_colb_pga(INPUT_COLB, INPUT_DT)
    if not pga_input_by_axis:
        print("⚠️ COLB.txt non trovato o non valido. PGA input sarà 0 (linee non mostrate).")

    # load excel
    excel_files = [EXCEL_BY_FLOORS[p] for p in FLOORS_TO_PLOT if p in EXCEL_BY_FLOORS]
    df_r, df_th = load_excel_bundle(excel_files)

    # filtro piani richiesti
    df_r = df_r[df_r["N_Piani"].isin(FLOORS_TO_PLOT)].copy()

    # PREPASS: limiti globali uguali per ogni tipo di plot
    lims = compute_global_limits(df_r, df_th, pga_input_by_axis)
    print(f"Limiti globali: TH ±{lims['th_y']:.3f} | PSD 0-{lims['psd_y']:.3e} | SA 0-{lims['sa_y']:.3f}")

    # plotting
    sa_cache = {}
    husid_rows = []

    groups = df_r[["N_Piani", "Nodo_ID", "Nodo_Tag", "Asse"]].drop_duplicates()

    for _, g in groups.iterrows():
        nP = int(g["N_Piani"])
        node_id = g["Nodo_ID"]
        node_tag = g["Nodo_Tag"]
        axis = g["Asse"]

        df_models = df_r[(df_r["N_Piani"] == nP) & (df_r["Nodo_ID"] == node_id) & (df_r["Asse"] == axis)].copy()
        if df_models.empty:
            continue

        fig1 = plot_triptych(nP, node_id, node_tag, axis, df_models, df_th, pga_input_by_axis, sa_cache, lims)
        fig1.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig1.savefig(os.path.join(DIR_COMPARE, f"CONFRONTO_{nP}P_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig1)

        fig2, rows = plot_husid(nP, node_id, node_tag, axis, df_models, df_th, pga_input_by_axis)
        husid_rows.extend(rows)
        fig2.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig2.savefig(os.path.join(DIR_HUSID, f"HUSID_{nP}P_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig2)

    # tabella husid/arias
    if husid_rows:
        pd.DataFrame(husid_rows).to_excel(os.path.join(DIR_TABLES, "TAB_HUSID_ARIAS_DURATE.xlsx"), index=False)

    # RIDUZIONI
    df_res = build_results_table(df_r, df_th)
    df_res.to_excel(os.path.join(DIR_TABLES, "TAB_PGA_SA_MAX_0_30s.xlsx"), index=False)

    df_red = compute_reductions(df_res, pga_input_by_axis)
    df_red.to_excel(os.path.join(DIR_TABLES, "TABELLA_RIDUZIONI.xlsx"), index=False)

    plot_reduction_histograms(df_red)

    # zip
    shutil.make_archive("POSTPROCESS_OUTPUT", "zip", OUTDIR)
    print("✅ Completato. Output in POSTPROCESS_OUTPUT/ e POSTPROCESS_OUTPUT.zip")


if __name__ == "__main__":
    main()
```

```python
!pip install pyrotd
import os
import re
import shutil
import tempfile # Added this import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.signal import welch
from scipy.integrate import cumulative_trapezoid

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e


# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81

FLOORS_TO_PLOT = [3,5,10,15]

# 1 file per N_piani (personalizza)
EXCEL_BY_FLOORS = {
    3:  "RISULTATI_3P.xlsx",
    5:  "RISULTATI_5P.xlsx",
    10: "RISULTATI_10P.xlsx",
    15: "RISULTATI_15P.xlsx",
}

# finestra di analisi comune
TMAX = 30.0

# input base
INPUT_COLB = "COLB.txt"
INPUT_DT = 0.01  # dt input COLB

# spettri (solo HNE/HNN)
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05
SA_XMAX = 3.0

# PSD (ricalcolata su 0–30 s)
PSD_FMAX = 50.0
WELCH_NPERSEG = 2048
PSD_MODE = "DB"  # "DB" oppure "LOG" oppure "LIN"
PSD_EPS = 1e-20  # per evitare log(0)

# Husid (durata Arias 5–95%)
HUSID_P1 = 0.05
HUSID_P2 = 0.95

# output dirs
OUTDIR = "POSTPROCESS_OUTPUT"
DIR_COMPARE = os.path.join(OUTDIR, "01_CONFRONTI_TH_PSD_SA")
DIR_HUSID = os.path.join(OUTDIR, "02_HUSID_ARIAS")
DIR_TABLES = os.path.join(OUTDIR, "03_TABELLE")
DIR_REDUCTIONS = os.path.join(OUTDIR, "04_RIDUZIONI_ISTOGRAMMI")


# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df


def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


def truncate_by_time(t: np.ndarray, x: np.ndarray, tmax: float):
    if len(t) == 0:
        return t, x
    mask = t <= tmax
    return t[mask], x[mask]


def extract_tis(model_name: str) -> float:
    m = re.search(r"(\d+\.\d+)", str(model_name))
    return float(m.group(1)) if m else 0.0


def classify_category(tipo_base: str, modello: str):
    """
    Categorie:
    - RIGIDA: base fissa
    - ISOLATA: elastomerici
    - SLITTE: elastomerici + slitte (friction)
    """
    tb = (tipo_base or "").upper()
    mn = (modello or "").upper()
    txt = tb if tb not in ("", "NAN", "NONE") else mn

    if ("RIGIDA" in txt) or ("FISSA" in txt) or ("FIXED" in txt):
        return "RIGIDA"
    if ("SLITT" in txt) or ("FRICTION" in txt):
        return "SLITTE"
    if ("ISOL" in txt) or ("ELAST" in txt):
        return "ISOLATA"
    return "ALTRO"


def style_for_category(cat: str, tis: float):
    if cat == "RIGIDA":
        return dict(color="black", lw=2.8, z=10, label="Base Fissa")
    if cat == "SLITTE":
        return dict(color="#D55E00", lw=1.8, z=5, label=f"Isolato+Slitte T={tis:.1f}s")
    if cat == "ISOLATA":
        return dict(color="#0072B2", lw=1.8, z=5, label=f"Isolato T={tis:.1f}s")
    return dict(color="gray", lw=1.2, z=1, label="Altro")


def sort_models_for_plot(df_models: pd.DataFrame) -> pd.DataFrame:
    order_map = {"RIGIDA": 0, "SLITTE": 1, "ISOLATA": 2, "ALTRO": 9}
    tmp = df_models.copy()
    tmp["__ord"] = tmp["Categoria"].map(order_map).fillna(9).astype(int)
    return tmp.sort_values(["__ord", "Tis", "Modello"]).drop(columns="__ord")


# ============================================================
# 2) INPUT: SOLO QUI PGA (input ground motion)
# ============================================================
def arias_curve(acc: np.ndarray, t: np.ndarray) -> np.ndarray:
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ


def read_colb_input_metrics(path: str, dt_input: float):
    """
    Input ground motion (solo questo è PGA):
      - PGA_input_by_axis: max|a(t)| su 0–30 s
      - Arias_input_by_axis: Ia(T) su 0–30 s
    COLB.txt: 3 colonne in g -> conversione a m/s².
    """
    if not os.path.exists(path):
        return {}, {}

    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    if df.shape[1] < 3:
        return {}, {}

    acc = {
        "HNE": df.iloc[:, 0].to_numpy(dtype=float) * G,
        "HNN": df.iloc[:, 1].to_numpy(dtype=float) * G,
        "HNZ": df.iloc[:, 2].to_numpy(dtype=float) * G,
    }

    pga_in, arias_in = {}, {}
    for ax, a in acc.items():
        t = np.arange(len(a)) * dt_input
        t, a = truncate_by_time(t, a, TMAX)

        pga_in[ax] = float(np.max(np.abs(a))) if len(a) else 0.0

        if len(a) >= 2:
            Ia = arias_curve(a, t)
            arias_in[ax] = float(Ia[-1])
        else:
            arias_in[ax] = 0.0

    return pga_in, arias_in


# ============================================================
# 3) HUSID (solo 0–30 s)
# ============================================================
def husid_metrics(Ia_curve_arr: np.ndarray, t: np.ndarray, p1=0.05, p2=0.95):
    Ia_tot = float(Ia_curve_arr[-1]) if len(Ia_curve_arr) else 0.0
    if Ia_tot <= 0:
        return dict(Ia_total=Ia_tot, t1=np.nan, t2=np.nan, duration=np.nan, husid=None)

    h = Ia_curve_arr / Ia_tot

    def _time_at(p):
        idx = np.searchsorted(h, p, side="left")
        if idx <= 0:
            return float(t[0])
        if idx >= len(t):
            return float(t[-1])
        t0, t1v = float(t[idx - 1]), float(t[idx])
        h0, h1 = float(h[idx - 1]), float(h[idx])
        if h1 == h0:
            return t1v
        return t0 + (p - h0) * (t1v - t0) / (h1 - h0)

    tp1 = _time_at(p1)
    tp2 = _time_at(p2)
    return dict(Ia_total=Ia_tot, t1=tp1, t2=tp2, duration=tp2 - tp1, husid=h)


# ============================================================
# 4) LETTURA EXCEL
# ============================================================
def load_excel_bundle(excel_files):
    list_r, list_th = [], []

    for f in excel_files:
        if not os.path.exists(f):
            print(f"⚠️ File non trovato: {f}")
            continue

        xls = pd.ExcelFile(f)
        if "Riepilogo" in xls.sheet_names:
            list_r.append(pd.read_excel(xls, sheet_name="Riepilogo"))
        if "TimeHistory" in xls.sheet_names:
            list_th.append(pd.read_excel(xls, sheet_name="TimeHistory"))

    if not list_r:
        raise ValueError("❌ Nessun foglio 'Riepilogo' caricato dagli Excel.")

    df_r = pd.concat(list_r, ignore_index=True)
    df_th = pd.concat(list_th, ignore_index=True) if list_th else pd.DataFrame()

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Tipo_Base", "Nodo_Tag", "Nodo_ID"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])

    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    if set(["Modello", "Nodo_ID", "Asse"]).issubset(df_r.columns):
        df_r = df_r.drop_duplicates(subset=["Modello", "Nodo_ID", "Asse"])

    if "Tipo_Base" not in df_r.columns:
        df_r["Tipo_Base"] = ""

    df_r["Tis"] = df_r["Modello"].apply(extract_tis)
    df_r["Categoria"] = df_r.apply(
        lambda r: classify_category(r.get("Tipo_Base", ""), r.get("Modello", "")),
        axis=1
    )

    return df_r, df_th


# ============================================================
# 5) PSD/SA (coerenti su 0–30 s)
# ============================================================
def compute_psd_welch(t: np.ndarray, a: np.ndarray):
    if len(t) < 4:
        return np.array([]), np.array([])
    dt = float(np.mean(np.diff(t)))
    fs = 1.0 / dt if dt > 0 else 0.0
    if fs <= 0:
        return np.array([]), np.array([])

    f, Pxx = welch(a, fs=fs, nperseg=min(WELCH_NPERSEG, len(a)))
    mask = f <= PSD_FMAX
    f = f[mask]
    Pxx = Pxx[mask]

    if PSD_MODE == "DB":
        return f, 10.0 * np.log10(Pxx + PSD_EPS)
    return f, Pxx


def compute_sa(t: np.ndarray, a: np.ndarray, cache: dict, key):
    if key in cache:
        return cache[key]
    if len(t) < 2:
        sa = np.zeros_like(PERIODS)
        cache[key] = sa
        return sa
    dt = float(np.mean(np.diff(t)))
    spec = pyrotd.calc_spec_accels(dt, a, OSC_FREQS, OSC_DAMPING)
    sa = np.asarray(spec.spec_accel, dtype=float)
    cache[key] = sa
    return sa


# ============================================================
# 6) PREPASS: LIMITI GLOBALI UGUALI
# ============================================================
def compute_global_limits(df_r, df_th, pga_input_by_axis):
    """
    Limiti globali (0–30 s):
    - TH: ±th_y (include anche PGA input)
    - PSD: y range (DB: [max-80, max+3], LOG/LIN: [0, max*1.15])
    - SA: [0, sa_y] (solo HNE/HNN)
    """
    max_acc = 0.0

    # include PGA input nel max TH
    for _, p in pga_input_by_axis.items():
        max_acc = max(max_acc, float(p))

    sa_cache = {}

    if PSD_MODE == "DB":
        max_psd = -np.inf
    else:
        max_psd = 0.0

    max_sa = 0.0

    keys = df_r[["N_Piani", "Nodo_ID", "Asse", "Modello"]].drop_duplicates()

    for _, k in keys.iterrows():
        mod = k["Modello"]
        node_id = k["Nodo_ID"]
        axis = k["Asse"]

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        if len(a):
            max_acc = max(max_acc, float(np.max(np.abs(a))))

        f, Pxx = compute_psd_welch(t, a)
        if len(Pxx):
            if PSD_MODE == "DB":
                max_psd = max(max_psd, float(np.max(Pxx)))
            else:
                max_psd = max(max_psd, float(np.max(Pxx)))

        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            if len(sa):
                max_sa = max(max_sa, float(np.max(sa)))

    lims = {}
    lims["th_y"] = 1.15 * max_acc if max_acc > 0 else 1.0

    if PSD_MODE == "DB":
        if np.isfinite(max_psd):
            lims["psd_ymax"] = max_psd + 3.0
            lims["psd_ymin"] = max_psd - 80.0
        else:
            lims["psd_ymax"] = 0.0
            lims["psd_ymin"] = -80.0
    else:
        lims["psd_ymax"] = 1.15 * max_psd if max_psd > 0 else 1.0
        lims["psd_ymin"] = 0.0

    lims["sa_y"] = 1.15 * max_sa if max_sa > 0 else 1.0
    return lims


# ============================================================
# 7) PLOT: TRIPTICO (TH/PSD/SA)
# ============================================================
def plot_triptych(nP, node_id, node_tag, axis, df_models, df_th, pga_input_by_axis, sa_cache, lims):
    fig, axs = plt.subplots(1, 3, figsize=(24, 7))
    fig.suptitle(f"{nP} Piani | {node_tag} ({node_id}) | {axis} | 0–30 s",
                 fontsize=16, fontweight="bold")

    # TH: solo linee ±PGA input (niente storia input)
    pga_in = float(pga_input_by_axis.get(axis, 0.0))
    if pga_in > 0:
        axs[0].axhline(+pga_in, color="gray", ls="--", lw=1.6, alpha=0.85, label="±PGA input")
        axs[0].axhline(-pga_in, color="gray", ls="--", lw=1.6, alpha=0.85)

    df_models = sort_models_for_plot(df_models)

    for _, r in df_models.iterrows():
        mod = r["Modello"]
        tis = float(r.get("Tis", 0.0))
        cat = r.get("Categoria", "ALTRO")
        st = style_for_category(cat, tis)

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        axs[0].plot(t, a, color=st["color"], lw=st["lw"], alpha=0.88, zorder=st["z"], label=st["label"])

        # PSD
        f, Pxx = compute_psd_welch(t, a)
        if len(f):
            axs[1].plot(f, Pxx, color=st["color"], lw=1.6, alpha=0.9, zorder=st["z"])

        # SA solo HNE/HNN
        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            axs[2].plot(PERIODS, sa, color=st["color"], lw=st["lw"], alpha=0.9, zorder=st["z"])

    # formatting (limiti globali)
    axs[0].set_title("Time History + ±PGA input", fontsize=13)
    axs[0].set_xlabel("Tempo [s]")
    axs[0].set_ylabel("Acc [m/s²]")
    axs[0].set_xlim(0, TMAX)
    axs[0].set_ylim(-lims["th_y"], lims["th_y"])
    axs[0].grid(True, ls=":", alpha=0.5)
    axs[0].legend(loc="upper right", fontsize=10)

    axs[1].set_title(f"PSD (Welch) su 0–30 s ({'dB' if PSD_MODE=='DB' else PSD_MODE})", fontsize=13)
    axs[1].set_xlabel("Frequenza [Hz]")
    axs[1].set_xlim(0, PSD_FMAX)
    axs[1].set_ylim(lims["psd_ymin"], lims["psd_ymax"])
    if PSD_MODE == "LOG":
        axs[1].set_yscale("log")
        axs[1].set_ylabel("PSD [unità²/Hz] (log)")
    elif PSD_MODE == "DB":
        axs[1].set_ylabel("PSD [dB]")
    else:
        axs[1].set_ylabel("PSD [unità²/Hz]")
    axs[1].grid(True, ls=":", alpha=0.5)

    if axis != "HNZ":
        axs[2].set_title("Spettro (5%) su 0–30 s", fontsize=13)
        axs[2].set_xlabel("Periodo [s]")
        axs[2].set_ylabel("SA [m/s²]")
        axs[2].set_xlim(0, SA_XMAX)
        axs[2].set_ylim(0, lims["sa_y"])
        axs[2].grid(True, ls=":", alpha=0.5)
    else:
        axs[2].set_visible(False)

    return fig


# ============================================================
# 8) PLOT: HUSID + TABELLA ARIAS/DURATE
# ============================================================
def plot_husid(nP, node_id, node_tag, axis, df_models, df_th):
    fig, ax = plt.subplots(figsize=(12, 7))
    fig.suptitle(f"Husid (Arias normalizzata) | {nP} Piani | {node_tag} ({node_id}) | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    rows = []
    df_models = sort_models_for_plot(df_models)

    for _, r in df_models.iterrows():
        mod = r["Modello"]
        tis = float(r.get("Tis", 0.0))
        cat = r.get("Categoria", "ALTRO")
        st = style_for_category(cat, tis)

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        Ia = arias_curve(a, t)
        m = husid_metrics(Ia, t, HUSID_P1, HUSID_P2)
        if m["husid"] is None:
            continue

        ax.plot(t, m["husid"], color=st["color"], lw=st["lw"], alpha=0.9,
                label=f"{st['label']} | D5-95={m['duration']:.2f}s")

        rows.append({
            "N_Piani": nP,
            "Nodo_ID": node_id,
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "Modello": mod,
            "Categoria": cat,
            "Tis": tis,
            "Arias_Output_0_30s": m["Ia_total"],
            "t5_s": m["t1"],
            "t95_s": m["t2"],
            "Durata_5_95_s": m["duration"],
        })

    ax.set_xlim(0, TMAX)
    ax.set_ylim(0, 1.02)
    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Husid = Ia(t)/Ia(T)")
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="lower right", fontsize=9, framealpha=0.95)

    return fig, rows


# ============================================================
# 9) RISULTATI: PFA (OUTPUT), SA, ARIAS
# ============================================================
def build_results_table(df_r, df_th):
    """
    Tabella (0–30 s) per ogni (N_Piani, Nodo_ID, Asse, Modello):
    - PFA_Output_0_30s = max|a_output| (PFA)
    - SA_Max_0_30s (solo HNE/HNN)
    - Arias_Output_0_30s
    """
    sa_cache = {}
    results = []

    groups = df_r[["N_Piani", "Nodo_ID", "Nodo_Tag", "Asse", "Modello", "Categoria", "Tis"]].drop_duplicates()

    for _, g in groups.iterrows():
        nP = int(g["N_Piani"])
        node_id = g["Nodo_ID"]
        node_tag = g["Nodo_Tag"]
        axis = g["Asse"]
        mod = g["Modello"]
        cat = g["Categoria"]
        tis = float(g["Tis"])

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        pfa = float(np.max(np.abs(a))) if len(a) else 0.0

        Ia_tot = 0.0
        if len(a) >= 2:
            Ia = arias_curve(a, t)
            Ia_tot = float(Ia[-1])

        sa_max = 0.0
        if axis != "HNZ" and len(a) >= 2:
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            sa_max = float(np.max(sa)) if len(sa) else 0.0

        results.append({
            "N_Piani": nP,
            "Nodo_ID": node_id,
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "Modello": mod,
            "Categoria": cat,
            "Tis": tis,
            "PFA_Output_0_30s": pfa,
            "SA_Max_0_30s": sa_max,
            "Arias_Output_0_30s": Ia_tot,
        })

    return pd.DataFrame(results)


# ============================================================
# 10) RIDUZIONI: PFA su HNE/HNN/HNZ; SA solo HNE/HNN
# ============================================================
def compute_reductions(df_res, pga_input_by_axis, arias_input_by_axis):
    """
    Riduzioni rispetto alla BASE FISSA (Categoria == RIGIDA) per ogni (N_Piani, Nodo_ID, Asse).
    - Riduzione_PFA_% sempre (anche HNZ)
    - Riduzione_SA_% solo se asse != HNZ (altrimenti NaN)
    Include anche PGA_input (solo input) e Arias_input.
    """
    out_rows = []

    grouped = df_res.groupby(["N_Piani", "Nodo_ID", "Asse"])
    for (nP, nID, ax), group in grouped:
        ref = group[group["Categoria"] == "RIGIDA"]
        if ref.empty:
            continue

        ref_pfa = float(ref.iloc[0]["PFA_Output_0_30s"])
        ref_sa = float(ref.iloc[0]["SA_Max_0_30s"])  # potrà essere 0 su HNZ

        for _, r in group.iterrows():
            if r["Categoria"] == "RIGIDA":
                continue

            pfa = float(r["PFA_Output_0_30s"])
            red_pfa = (1.0 - pfa / ref_pfa) * 100.0 if ref_pfa > 0 else 0.0

            red_sa = np.nan
            if ax != "HNZ":
                sa = float(r["SA_Max_0_30s"])
                red_sa = (1.0 - sa / ref_sa) * 100.0 if ref_sa > 0 else 0.0

            tis = float(r["Tis"])
            strategia = ("Slitte " if r["Categoria"] == "SLITTE" else "Isolato ") + f"{tis:.1f}s"

            out_rows.append({
                "N_Piani": int(nP),
                "Nodo_ID": nID,
                "Asse": ax,
                "Categoria": r["Categoria"],
                "Tis": tis,
                "Strategia": strategia,
                "Modello": r["Modello"],

                "PGA_Input_0_30s": float(pga_input_by_axis.get(ax, 0.0)),   # SOLO input
                "Arias_Input_0_30s": float(arias_input_by_axis.get(ax, 0.0)),

                "PFA_Output_0_30s": pfa,
                "Riduzione_PFA_%": round(float(red_pfa), 2),

                "SA_Max_0_30s": float(r["SA_Max_0_30s"]),
                "Riduzione_SA_%": (round(float(red_sa), 2) if np.isfinite(red_sa) else np.nan),

                "Arias_Output_0_30s": float(r["Arias_Output_0_30s"]),
            })

    return pd.DataFrame(out_rows)


def plot_reduction_histograms(df_red):
    """
    Istogrammi per ciascun N_piani:
    - Riduzione PFA (%) per HNE/HNN/HNZ
    - Riduzione SA (%) per HNE/HNN
    Le barre sono medie su tutti i nodi (come prassi nel tuo script).
    """
    if df_red.empty:
        return

    os.makedirs(DIR_REDUCTIONS, exist_ok=True)

    # ordina strategie per categoria e Tis
    df_plot = df_red.copy()
    cat_order = {"SLITTE": 0, "ISOLATA": 1, "ALTRO": 9}
    df_plot["__catord"] = df_plot["Categoria"].map(cat_order).fillna(9).astype(int)
    df_plot = df_plot.sort_values(["__catord", "Tis"]).drop(columns="__catord")

    for nP in sorted(df_plot["N_Piani"].unique()):
        sub = df_plot[df_plot["N_Piani"] == nP].copy()
        if sub.empty:
            continue

        # media su nodi
        summary = sub.groupby(["Asse", "Strategia"])[["Riduzione_PFA_%", "Riduzione_SA_%"]].mean().reset_index()

        strategies = list(dict.fromkeys(sub.sort_values(["Categoria", "Tis"])["Strategia"].tolist()))
        x = np.arange(len(strategies))

        fig, axs = plt.subplots(1, 2, figsize=(18, 9))
        fig.suptitle(f"Efficienza Isolamento (0–30 s) – {nP} Piani", fontsize=18, fontweight="bold")

        # --- PFA: HNE/HNN/HNZ ---
        assi_pfa = ["HNE", "HNN", "HNZ"]
        width = 0.25
        for i, ax_code in enumerate(assi_pfa):
            vals = []
            for s in strategies:
                row = summary[(summary["Asse"] == ax_code) & (summary["Strategia"] == s)]
                vals.append(float(row["Riduzione_PFA_%"].values[0]) if not row.empty else 0.0)
            offset = (i - 1) * width
            rects = axs[0].bar(x + offset, vals, width, label=ax_code, alpha=0.85, edgecolor="black")
            axs[0].bar_label(rects, fmt="%.0f%%", padding=3, fontsize=9, rotation=90)

        axs[0].set_title("Riduzione PFA output vs Base Fissa", fontsize=13)
        axs[0].set_ylabel("Riduzione [%]", fontsize=12)
        axs[0].set_xticks(x)
        axs[0].set_xticklabels(strategies, rotation=45, ha="right", fontsize=10)
        axs[0].set_ylim(0, 100)
        axs[0].grid(axis="y", ls="--", alpha=0.5)
        axs[0].legend(loc="lower left")

        # --- SA: HNE/HNN ---
        assi_sa = ["HNE", "HNN"]
        width2 = 0.30
        for i, ax_code in enumerate(assi_sa):
            vals = []
            for s in strategies:
                row = summary[(summary["Asse"] == ax_code) & (summary["Strategia"] == s)]
                v = float(row["Riduzione_SA_%"].values[0]) if not row.empty else 0.0
                vals.append(v if np.isfinite(v) else 0.0)
            offset = (i - 0.5) * width2
            rects = axs[1].bar(x + offset, vals, width2, label=ax_code, alpha=0.85, edgecolor="black")
            axs[1].bar_label(rects, fmt="%.0f%%", padding=3, fontsize=9, rotation=90)

        axs[1].set_title("Riduzione SA max output vs Base Fissa (HNE/HNN)", fontsize=13)
        axs[1].set_ylabel("Riduzione [%]", fontsize=12)
        axs[1].set_xticks(x)
        axs[1].set_xticklabels(strategies, rotation=45, ha="right", fontsize=10)
        axs[1].set_ylim(0, 100)
        axs[1].grid(axis="y", ls="--", alpha=0.5)
        axs[1].legend(loc="lower left")

        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_REDUCTIONS, f"ISTOGRAMMI_RIDUZIONI_{nP}P.png"), dpi=150)
        plt.close(fig)


# ============================================================
# 11) MAIN
# ============================================================
def main():
    # reset output
    if os.path.exists(OUTDIR):
        shutil.rmtree(OUTDIR)
    os.makedirs(DIR_COMPARE, exist_ok=True)
    os.makedirs(DIR_HUSID, exist_ok=True)
    os.makedirs(DIR_TABLES, exist_ok=True)
    os.makedirs(DIR_REDUCTIONS, exist_ok=True)

    # INPUT: PGA + Arias (solo input)
    PGA_input_by_axis, Arias_input_by_axis = read_colb_input_metrics(INPUT_COLB, INPUT_DT)
    if not PGA_input_by_axis:
        print("⚠️ COLB.txt non trovato o non valido. PGA input sarà 0 e linee non verranno mostrate.")

    # load excel (solo piani richiesti)
    excel_files = [EXCEL_BY_FLOORS[p] for p in FLOORS_TO_PLOT if p in EXCEL_BY_FLOORS]
    df_r, df_th = load_excel_bundle(excel_files)
    df_r = df_r[df_r["N_Piani"].isin(FLOORS_TO_PLOT)].copy()

    # limiti globali uguali
    lims = compute_global_limits(df_r, df_th, PGA_input_by_axis)
    print(
        f"Limiti globali: TH ±{lims['th_y']:.3f} | PSD [{lims['psd_ymin']:.2f}, {lims['psd_ymax']:.2f}]"
        f" | SA [0, {lims['sa_y']:.3f}]"
    )

    # plot per gruppi
    sa_cache = {}
    husid_rows = []

    groups = df_r[["N_Piani", "Nodo_ID", "Nodo_Tag", "Asse"]].drop_duplicates()

    for _, g in groups.iterrows():
        nP = int(g["N_Piani"])
        node_id = g["Nodo_ID"]
        node_tag = g["Nodo_Tag"]
        axis = g["Asse"]

        df_models = df_r[(df_r["N_Piani"] == nP) & (df_r["Nodo_ID"] == node_id) & (df_r["Asse"] == axis)].copy()
        if df_models.empty:
            continue

        # triptico
        fig1 = plot_triptych(
            nP, node_id, node_tag, axis,
            df_models, df_th,
            PGA_input_by_axis,
            sa_cache, lims
        )
        fig1.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig1.savefig(os.path.join(DIR_COMPARE, f"CONFRONTO_{nP}P_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig1)

        # husid
        fig2, rows = plot_husid(nP, node_id, node_tag, axis, df_models, df_th)
        husid_rows.extend(rows)
        fig2.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig2.savefig(os.path.join(DIR_HUSID, f"HUSID_{nP}P_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig2)

    # Tabelle Arias/Husid
    if husid_rows:
        df_husid = pd.DataFrame(husid_rows)
        df_husid.to_excel(os.path.join(DIR_TABLES, "TAB_HUSID_ARIAS_DURATE_0_30s.xlsx"), index=False)

    # Tabella input (PGA + Arias)
    df_in = pd.DataFrame([
        {"Asse": ax, "PGA_Input_0_30s": float(PGA_input_by_axis.get(ax, 0.0)), "Arias_Input_0_30s": float(Arias_input_by_axis.get(ax, 0.0))}
        for ax in ["HNE", "HNN", "HNZ"]
    ])
    df_in.to_excel(os.path.join(DIR_TABLES, "TAB_INPUT_PGA_ARIAS_0_30s.xlsx"), index=False)

    # Risultati output: PFA + SA + Arias
    df_res = build_results_table(df_r, df_th)
    df_res.to_excel(os.path.join(DIR_TABLES, "TAB_OUTPUT_PFA_SA_ARIAS_0_30s.xlsx"), index=False)

    # Tabella dedicata intensità Arias output
    cols_arias = ["N_Piani","Nodo_ID","Nodo_Tag","Asse","Modello","Categoria","Tis","Arias_Output_0_30s"]
    df_res[cols_arias].to_excel(os.path.join(DIR_TABLES, "TAB_ARIAS_OUTPUT_INTENSITY_0_30s.xlsx"), index=False)

    # Riduzioni + istogrammi
    df_red = compute_reductions(df_res, PGA_input_by_axis, Arias_input_by_axis)
    df_red.to_excel(os.path.join(DIR_TABLES, "TABELLA_RIDUZIONI_PFA_SA_ARIAS_0_30s.xlsx"), index=False)
    plot_reduction_histograms(df_red)

    # Nuova sezione per zippare tutto
    with tempfile.TemporaryDirectory() as tmp_dir:
        # Copia la directory di output del post-processing
        if os.path.exists(OUTDIR):
            shutil.copytree(OUTDIR, os.path.join(tmp_dir, OUTDIR))
        else:
            print(f"⚠️ La cartella {OUTDIR} non trovata, non sarà inclusa nello ZIP.")

        # Copia il file COLB.txt
        if os.path.exists(INPUT_COLB):
            shutil.copy(INPUT_COLB, os.path.join(tmp_dir, INPUT_COLB))
        else:
            print(f"⚠️ {INPUT_COLB} non trovato, non sarà incluso nello ZIP.")

        # Copia i file RISULTATI_*.xlsx
        for nP, excel_name in EXCEL_BY_FLOORS.items():
            if os.path.exists(excel_name):
                shutil.copy(excel_name, os.path.join(tmp_dir, excel_name))
            else:
                print(f"⚠️ {excel_name} (per {nP} Piani) non trovato, non sarà incluso nello ZIP.")

        # Crea l'archivio ZIP finale
        final_zip_name = "FULL_REPORT"
        shutil.make_archive(final_zip_name, "zip", tmp_dir)

    print("✅ Completato. Tutti i risultati sono stati archiviati in FULL_REPORT.zip")


if __name__ == "__main__":
    main()
```

```python
# ============================================================
# SCRIPT DI SINTESI "PAPER-READY" – ROBUSTO A STRATEGIE MANCANTI
# (da eseguire in Colab DOPO lo script precedente)
# ============================================================
# Questo script NON assume che tutte le strategie (es. "Isolato 2.0s") esistano
# per tutti i numeri di piani. Per ciascun N_Piani usa SOLO le strategie presenti
# in quel sottoinsieme, quindi non si blocca per 15 piani senza 2.0 s.
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# PATHS
# -----------------------------
BASE_DIR = "POSTPROCESS_OUTPUT"
TABLE_DIR = os.path.join(BASE_DIR, "03_TABELLE")
OUT_DIR = os.path.join(BASE_DIR, "05_SINTESI_PAPER")
os.makedirs(OUT_DIR, exist_ok=True)

FILE_RES = os.path.join(TABLE_DIR, "TAB_OUTPUT_PFA_SA_ARIAS_0_30s.xlsx")
FILE_RED = os.path.join(TABLE_DIR, "TABELLA_RIDUZIONI_PFA_SA_ARIAS_0_30s.xlsx")
FILE_HUS = os.path.join(TABLE_DIR, "TAB_HUSID_ARIAS_DURATE_0_30s.xlsx")  # optional

AXES_ALL = ["HNE", "HNN", "HNZ"]
AXES_HOR = ["HNE", "HNN"]

# -----------------------------
# HELPERS
# -----------------------------
def finite(x):
    x = np.asarray(x, dtype=float)
    return x[np.isfinite(x)]

def q(x, p):
    x = finite(x)
    if x.size == 0:
        return np.nan
    return float(np.percentile(x, p))

def stats_iqr(x):
    x = finite(x)
    if x.size == 0:
        return dict(n=0, mean=np.nan, median=np.nan, q1=np.nan, q3=np.nan)
    return dict(
        n=int(x.size),
        mean=float(np.mean(x)),
        median=float(np.median(x)),
        q1=q(x, 25),
        q3=q(x, 75)
    )

def ensure_cols(df, required, name):
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Nel file {name} mancano colonne: {missing}")

def clean_strategy(s):
    if pd.isna(s):
        return "N/A"
    return str(s).strip()

def make_strategy(df):
    """
    Costruisce una colonna Strategia coerente con il naming:
      - "Slitte {Tis:.1f}s" se Categoria==SLITTE
      - "Isolato {Tis:.1f}s" se Categoria==ISOLATA
      - altrimenti Categoria
    """
    out = df.copy()
    out["Strategia"] = out.apply(
        lambda r: (("Slitte " if r["Categoria"] == "SLITTE" else "Isolato ") + f"{float(r['Tis']):.1f}s")
        if r["Categoria"] in ("SLITTE", "ISOLATA") else str(r["Categoria"]),
        axis=1
    )
    out["Strategia"] = out["Strategia"].apply(clean_strategy)
    return out

def strategy_order_floor(df_floor):
    """
    Ordina le strategie presenti SOLO nel sottoinsieme (N_Piani) corrente.
    Se una strategia manca (es. 2.0s nei 15 piani), non viene inclusa.
    """
    if df_floor.empty:
        return []
    tmp = df_floor[["Strategia", "Categoria", "Tis"]].drop_duplicates().copy()
    cat_order = {"SLITTE": 0, "ISOLATA": 1, "ALTRO": 9}
    tmp["__c"] = tmp["Categoria"].map(cat_order).fillna(9).astype(int)
    tmp = tmp.sort_values(["__c", "Tis", "Strategia"])
    return list(tmp["Strategia"].tolist())

def safe_boxplot(ax, data_list, labels):
    # boxplot robusto: se un gruppo è vuoto lo salta (e aggiorna labels coerentemente)
    d2, l2 = [], []
    for d, l in zip(data_list, labels):
        d = finite(d)
        if d.size:
            d2.append(d)
            l2.append(l)
    if not d2:
        ax.text(0.5, 0.5, "Nessun dato", ha="center", va="center")
        return
    ax.boxplot(d2, labels=l2, showfliers=False)
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
        tick.set_ha("right")

def safe_dot_iqr(ax, y_pos, mean, q1v, q3v):
    if np.isfinite(q1v) and np.isfinite(q3v):
        ax.plot([q1v, q3v], [y_pos, y_pos], lw=3, alpha=0.9)
    if np.isfinite(mean):
        ax.scatter([mean], [y_pos], s=60)

# -----------------------------
# LOAD
# -----------------------------
if not os.path.exists(FILE_RES):
    raise FileNotFoundError(FILE_RES)
if not os.path.exists(FILE_RED):
    raise FileNotFoundError(FILE_RED)

df_res = pd.read_excel(FILE_RES)
df_red = pd.read_excel(FILE_RED)

df_hus = None
if os.path.exists(FILE_HUS):
    df_hus = pd.read_excel(FILE_HUS)

ensure_cols(df_res, ["N_Piani","Asse","Categoria","Tis","PFA_Output_0_30s","Arias_Output_0_30s","SA_Max_0_30s"], "TAB_OUTPUT_PFA_SA_ARIAS_0_30s.xlsx")
ensure_cols(df_red, ["N_Piani","Asse","Categoria","Tis","Strategia","Riduzione_PFA_%","Riduzione_SA_%"], "TABELLA_RIDUZIONI_PFA_SA_ARIAS_0_30s.xlsx")

df_red["Strategia"] = df_red["Strategia"].apply(clean_strategy)
# ricostruisco Strategia anche in df_res per consistenza
df_res = make_strategy(df_res)

if df_hus is not None:
    ensure_cols(df_hus, ["N_Piani","Asse","Categoria","Tis","Durata_5_95_s"], "TAB_HUSID_ARIAS_DURATE_0_30s.xlsx")
    df_hus = make_strategy(df_hus)

# -----------------------------
# 1) BOX PLOTS: Riduzioni (%) per asse
# -----------------------------
def plot_box_reductions(df_red, nP):
    sub = df_red[df_red["N_Piani"] == nP].copy()
    sub = sub[sub["Categoria"].isin(["SLITTE","ISOLATA"])].copy()
    if sub.empty:
        return

    order = strategy_order_floor(sub)

    # PFA reductions: HNE/HNN/HNZ
    fig, axs = plt.subplots(1, 3, figsize=(20, 6), sharey=True)
    fig.suptitle(f"{int(nP)} piani – Riduzione PFA (%) vs base fissa (distribuzione su nodi)", fontsize=16, fontweight="bold")

    for i, ax_code in enumerate(AXES_ALL):
        ax = axs[i]
        data = [sub[(sub["Asse"] == ax_code) & (sub["Strategia"] == s)]["Riduzione_PFA_%"] for s in order]
        safe_boxplot(ax, data, order)
        ax.axhline(0, lw=1, ls="--", alpha=0.6)
        ax.set_title(ax_code)
        ax.set_xlabel("Strategia")
        ax.grid(True, ls=":", alpha=0.4)

    axs[0].set_ylabel("Riduzione PFA [%]")
    fig.tight_layout(rect=[0, 0.02, 1, 0.93])
    fig.savefig(os.path.join(OUT_DIR, f"{int(nP)}P_BOX_RIDUZIONE_PFA.png"), dpi=220)
    plt.close(fig)

    # SA reductions: HNE/HNN
    fig, axs = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
    fig.suptitle(f"{int(nP)} piani – Riduzione SAmax (%) vs base fissa (distribuzione su nodi)", fontsize=16, fontweight="bold")

    for i, ax_code in enumerate(AXES_HOR):
        ax = axs[i]
        data = [sub[(sub["Asse"] == ax_code) & (sub["Strategia"] == s)]["Riduzione_SA_%"] for s in order]
        safe_boxplot(ax, data, order)
        ax.axhline(0, lw=1, ls="--", alpha=0.6)
        ax.set_title(ax_code)
        ax.set_xlabel("Strategia")
        ax.grid(True, ls=":", alpha=0.4)

    axs[0].set_ylabel("Riduzione SA [%]")
    fig.tight_layout(rect=[0, 0.02, 1, 0.93])
    fig.savefig(os.path.join(OUT_DIR, f"{int(nP)}P_BOX_RIDUZIONE_SA.png"), dpi=220)
    plt.close(fig)

# -----------------------------
# 2) Trade-off plot: Riduzione PFA vs Riduzione SA (HNE/HNN)
# -----------------------------
def plot_tradeoff(df_red, nP):
    sub = df_red[(df_red["N_Piani"] == nP) & (df_red["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
    sub = sub[sub["Asse"].isin(AXES_HOR)].copy()
    if sub.empty:
        return

    order = strategy_order_floor(sub)

    fig, ax = plt.subplots(figsize=(9, 7))
    fig.suptitle(f"{int(nP)} piani – Trade-off: Riduzione PFA vs Riduzione SA (HNE/HNN)", fontsize=14, fontweight="bold")

    for s in order:
        ss = sub[sub["Strategia"] == s]
        x = finite(ss["Riduzione_PFA_%"])
        y = finite(ss["Riduzione_SA_%"])
        if x.size == 0 or y.size == 0:
            continue

        ax.scatter(x, y, s=22, alpha=0.55, label=s)

        # rettangolo IQR (robusto)
        x1, x3 = q(x, 25), q(x, 75)
        y1, y3 = q(y, 25), q(y, 75)
        if np.isfinite(x1) and np.isfinite(x3) and np.isfinite(y1) and np.isfinite(y3):
            ax.plot([x1, x3, x3, x1, x1], [y1, y1, y3, y3, y1], lw=1.3, alpha=0.9)

        ax.scatter([q(x, 50)], [q(y, 50)], s=80, marker="x", linewidths=2)

    ax.axhline(0, lw=1, ls="--", alpha=0.6)
    ax.axvline(0, lw=1, ls="--", alpha=0.6)
    ax.set_xlabel("Riduzione PFA [%]")
    ax.set_ylabel("Riduzione SA [%]")
    ax.grid(True, ls=":", alpha=0.4)
    ax.legend(fontsize=9, loc="lower right", framealpha=0.95)
    fig.tight_layout(rect=[0, 0.02, 1, 0.93])
    fig.savefig(os.path.join(OUT_DIR, f"{int(nP)}P_TRADEOFF_PFA_SA.png"), dpi=240)
    plt.close(fig)

# -----------------------------
# 3) Summary dot-plot: mean ± IQR (valori assoluti)
# -----------------------------
def dotplot_mean_iqr(df, metric, axes_list, title, xlabel, fname, nP):
    sub = df[(df["N_Piani"] == nP) & (df["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
    if sub.empty:
        return

    order = strategy_order_floor(sub)

    fig, axs = plt.subplots(1, len(axes_list), figsize=(6*len(axes_list), 6), sharey=True)
    if len(axes_list) == 1:
        axs = [axs]

    fig.suptitle(f"{int(nP)} piani – {title} (mean ± IQR su nodi)", fontsize=15, fontweight="bold")

    for i, ax_code in enumerate(axes_list):
        ax = axs[i]
        y_pos = np.arange(len(order))
        ax.set_yticks(y_pos)
        ax.set_yticklabels(order, fontsize=10)
        ax.set_xlabel(xlabel)
        ax.set_title(ax_code)
        ax.grid(True, ls=":", alpha=0.4)

        for j, s in enumerate(order):
            x = sub[(sub["Asse"] == ax_code) & (sub["Strategia"] == s)][metric]
            st = stats_iqr(x)
            safe_dot_iqr(ax, j, st["mean"], st["q1"], st["q3"])

    fig.tight_layout(rect=[0, 0.02, 1, 0.93])
    fig.savefig(os.path.join(OUT_DIR, f"{int(nP)}P_{fname}.png"), dpi=240)
    plt.close(fig)

# -----------------------------
# 4) Durata D5–95: mean ± IQR (se disponibile)
# -----------------------------
def plot_duration_dot(df_hus, nP):
    if df_hus is None:
        return
    sub = df_hus[(df_hus["N_Piani"] == nP) & (df_hus["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
    if sub.empty:
        return
    dotplot_mean_iqr(
        df=sub,
        metric="Durata_5_95_s",
        axes_list=AXES_ALL,
        title="Durata energetica D5–95",
        xlabel="Durata [s]",
        fname="DOT_DURATA_D5_95",
        nP=nP
    )

# -----------------------------
# 5) Tabella finale “paper-ready” + testo sintesi (robusti)
# -----------------------------
def build_paper_summary_table(df_res, df_red, df_hus):
    rows = []

    def add(kind, nP, ax, s, dct, unit):
        rows.append({
            "Tipo": kind,
            "N_Piani": int(nP),
            "Asse": ax,
            "Strategia": s,
            "n": dct["n"],
            "mean": dct["mean"],
            "median": dct["median"],
            "q1": dct["q1"],
            "q3": dct["q3"],
            "Unita": unit,
        })

    floors = sorted(df_red["N_Piani"].dropna().unique())
    for nP in floors:
        # strategie presenti SOLO in questo piano
        subr = df_red[(df_red["N_Piani"] == nP) & (df_red["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
        if subr.empty:
            continue
        order = strategy_order_floor(subr)

        # riduzioni
        for ax_code in AXES_ALL:
            for s in order:
                add("Riduzione_PFA", nP, ax_code, s,
                    stats_iqr(subr[(subr["Asse"]==ax_code) & (subr["Strategia"]==s)]["Riduzione_PFA_%"]), "%")
        for ax_code in AXES_HOR:
            for s in order:
                add("Riduzione_SA", nP, ax_code, s,
                    stats_iqr(subr[(subr["Asse"]==ax_code) & (subr["Strategia"]==s)]["Riduzione_SA_%"]), "%")

        # assoluti
        suba = df_res[(df_res["N_Piani"] == nP) & (df_res["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
        if not suba.empty:
            for ax_code in AXES_ALL:
                for s in order:
                    add("PFA_output", nP, ax_code, s,
                        stats_iqr(suba[(suba["Asse"]==ax_code) & (suba["Strategia"]==s)]["PFA_Output_0_30s"]), "m/s^2")
                    add("Arias_output", nP, ax_code, s,
                        stats_iqr(suba[(suba["Asse"]==ax_code) & (suba["Strategia"]==s)]["Arias_Output_0_30s"]), "-")
            for ax_code in AXES_HOR:
                for s in order:
                    add("SAmax_output", nP, ax_code, s,
                        stats_iqr(suba[(suba["Asse"]==ax_code) & (suba["Strategia"]==s)]["SA_Max_0_30s"]), "m/s^2")

        # durate
        if df_hus is not None:
            subd = df_hus[(df_hus["N_Piani"] == nP) & (df_hus["Categoria"].isin(["SLITTE","ISOLATA"]))].copy()
            if not subd.empty:
                for ax_code in AXES_ALL:
                    for s in order:
                        add("Durata_D5_95", nP, ax_code, s,
                            stats_iqr(subd[(subd["Asse"]==ax_code) & (subd["Strategia"]==s)]["Durata_5_95_s"]), "s")

    return pd.DataFrame(rows)

def write_paper_paragraph(df_sum, out_txt):
    def fmt(x, nd=0):
        if x is None or (isinstance(x, float) and not np.isfinite(x)):
            return "n.d."
        return f"{x:.{nd}f}"

    lines = []
    lines.append("Sintesi risultati (0–30 s) – statistiche robuste (mediana e IQR) su tutti i nodi.\n")

    floors = sorted(df_sum["N_Piani"].unique())
    for nP in floors:
        lines.append(f"{int(nP)} piani:")
        sub = df_sum[(df_sum["N_Piani"]==nP) & (df_sum["Tipo"].isin(["Riduzione_PFA","Riduzione_SA"]))].copy()
        if sub.empty:
            lines.append("  Dati riduzioni non disponibili.")
            continue

        strategies = list(dict.fromkeys(sub["Strategia"].tolist()))
        for s in strategies:
            def pull(kind, axis):
                row = sub[(sub["Tipo"]==kind) & (sub["Asse"]==axis) & (sub["Strategia"]==s)]
                if row.empty:
                    return (np.nan, np.nan, np.nan)
                r = row.iloc[0]
                return (float(r["median"]), float(r["q1"]), float(r["q3"]))

            m_hne, q1_hne, q3_hne = pull("Riduzione_PFA", "HNE")
            m_hnn, q1_hnn, q3_hnn = pull("Riduzione_PFA", "HNN")
            m_hnz, q1_hnz, q3_hnz = pull("Riduzione_PFA", "HNZ")

            m_sa_hne, q1_sa_hne, q3_sa_hne = pull("Riduzione_SA", "HNE")
            m_sa_hnn, q1_sa_hnn, q3_sa_hnn = pull("Riduzione_SA", "HNN")

            lines.append(
                f"  - {s}: "
                f"Riduzione PFA mediana HNE={fmt(m_hne)}% (IQR {fmt(q1_hne)}–{fmt(q3_hne)}), "
                f"HNN={fmt(m_hnn)}% (IQR {fmt(q1_hnn)}–{fmt(q3_hnn)}), "
                f"HNZ={fmt(m_hnz)}% (IQR {fmt(q1_hnz)}–{fmt(q3_hnz)}); "
                f"Riduzione SA mediana HNE={fmt(m_sa_hne)}% (IQR {fmt(q1_sa_hne)}–{fmt(q3_sa_hne)}), "
                f"HNN={fmt(m_sa_hnn)}% (IQR {fmt(q1_sa_hnn)}–{fmt(q3_sa_hnn)})."
            )

    with open(out_txt, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

# -----------------------------
# RUN
# -----------------------------
floors = sorted(df_red["N_Piani"].dropna().unique())

for nP in floors:
    plot_box_reductions(df_red, nP)
    plot_tradeoff(df_red, nP)

    dotplot_mean_iqr(df_res, "PFA_Output_0_30s", AXES_ALL, "PFA output", "PFA [m/s²]", "DOT_PFA_OUTPUT", nP)
    dotplot_mean_iqr(df_res, "Arias_Output_0_30s", AXES_ALL, "Arias output (Ia)", "Ia [-]", "DOT_ARIAS_OUTPUT", nP)
    dotplot_mean_iqr(df_res, "SA_Max_0_30s", AXES_HOR, "SA max output", "SA max [m/s²]", "DOT_SAmax_OUTPUT", nP)
    plot_duration_dot(df_hus, nP)

df_sum = build_paper_summary_table(df_res, df_red, df_hus)
df_sum.to_excel(os.path.join(OUT_DIR, "SUMMARY_PAPER_READY.xlsx"), index=False)
write_paper_paragraph(df_sum, os.path.join(OUT_DIR, "PARAGRAFO_SINTESI_PAPER_READY.txt"))

print("OK: grafici paper-ready robusti a strategie mancanti in:", OUT_DIR)
print(" - SUMMARY_PAPER_READY.xlsx")
print(" - PARAGRAFO_SINTESI_PAPER_READY.txt")
print(" - figure *.png (box, trade-off, dot-plot) per ogni N_piani (solo strategie presenti)")
```

```python
# ============================================================
# BLOCCO COLAB – PLOT "DECISION SUPPORT" PER BRADISISMO
# ============================================================
# Obiettivo (stile research/decision-support):
# 1) Dimostrare in modo chiaro che l’isolamento è efficace nel ridurre la domanda
#    (proxy: Riduzione PFA e SA) in un contesto di eccitazione a contenuto low-frequency
#    tipico del bradisismo (argomentazione basata su metriche di risposta).
# 2) Rendere esplicito, per ogni N_Piani, quale tipologia di isolamento scegliere
#    (Isolato vs Isolato+Slitte) e quale periodo Tis è "migliore" secondo una funzione-obiettivo.
#
# NOTE IMPORTANTI:
# - Questo blocco NON ricostruisce la fisica del bradisismo (servirebbero segnali lenti e drift);
#   usa però indicatori di mitigazione della domanda dinamica (PFA/SA) come evidenza quantitativa.
# - Se vuoi una dimostrazione "bradisismo-specifica" rigorosa, bisogna includere anche
#   drift/IDR, spostamenti, o filtrare la risposta in banda bassa (es. < 1 Hz). Qui ci basiamo
#   sulle riduzioni già calcolate (0–30 s).
#
# Output: POSTPROCESS_OUTPUT/07_BRADISISMO_DECISION
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = "POSTPROCESS_OUTPUT"
TABLE_DIR = os.path.join(BASE_DIR, "03_TABELLE")
OUT_DIR = os.path.join(BASE_DIR, "07_BRADISISMO_DECISION")
os.makedirs(OUT_DIR, exist_ok=True)

FILE_RED = os.path.join(TABLE_DIR, "TABELLA_RIDUZIONI_PFA_SA_ARIAS_0_30s.xlsx")
FILE_RES = os.path.join(TABLE_DIR, "TAB_OUTPUT_PFA_SA_ARIAS_0_30s.xlsx")
if not os.path.exists(FILE_RED):
    raise FileNotFoundError(FILE_RED)
if not os.path.exists(FILE_RES):
    raise FileNotFoundError(FILE_RES)

df_red = pd.read_excel(FILE_RED)
df_res = pd.read_excel(FILE_RES)

AXES_ALL = ["HNE", "HNN", "HNZ"]
AXES_HOR = ["HNE", "HNN"]

# -----------------------------
# HELPERS
# -----------------------------
def finite(x):
    x = np.asarray(x, dtype=float)
    return x[np.isfinite(x)]

def pct(x, p):
    x = finite(x)
    if x.size == 0:
        return np.nan
    return float(np.percentile(x, p))

def robust_stats(x):
    x = finite(x)
    if x.size == 0:
        return dict(n=0, med=np.nan, q1=np.nan, q3=np.nan, mean=np.nan)
    return dict(n=int(x.size), med=pct(x,50), q1=pct(x,25), q3=pct(x,75), mean=float(np.mean(x)))

def clean_strategy(s):
    if pd.isna(s):
        return "N/A"
    return str(s).strip()

def ensure_cols(df, required):
    miss = [c for c in required if c not in df.columns]
    if miss:
        raise ValueError(f"Colonne mancanti: {miss}")

def category_label(cat):
    if cat == "ISOLATA":
        return "Isolato (solo elastomeri)"
    if cat == "SLITTE":
        return "Isolato + Slitte"
    if cat == "RIGIDA":
        return "Base fissa"
    return str(cat)

ensure_cols(df_red, ["N_Piani","Asse","Categoria","Tis","Strategia","Riduzione_PFA_%","Riduzione_SA_%"])
ensure_cols(df_res, ["N_Piani","Asse","Categoria","Tis","PFA_Output_0_30s","SA_Max_0_30s","Arias_Output_0_30s"])

df_red["Strategia"] = df_red["Strategia"].apply(clean_strategy)

# Considero solo isolamenti nel decision (Isolata/Slitte); base fissa è riferimento già incorporato nelle riduzioni.
df_iso = df_red[df_red["Categoria"].isin(["ISOLATA","SLITTE"])].copy()

floors = sorted(df_iso["N_Piani"].dropna().unique())

# -----------------------------
# 0) COSTRUISCI FUNZIONE-OBIETTIVO (score)
# -----------------------------
# In ottica "bradisismo" (contenuto a bassa frequenza, ma potenzialmente lunghi periodi),
# assumiamo come scelta robusta:
# - massimizzare riduzione PFA orizzontale (media HNE/HNN)
# - massimizzare riduzione SA orizzontale (media HNE/HNN)
# - penalizzare riduzione verticale troppo bassa (HNZ) in modo lieve (opzionale)
#
# Score = wPFA*median(PFAred_Hmean) + wSA*median(SAred_Hmean) + wV*median(PFAred_HNZ)
# Default pesi: PFA 0.55, SA 0.35, Vertical 0.10
wPFA, wSA, wV = 0.55, 0.35, 0.10

def compute_strategy_score(df_floor, strategy):
    """
    Restituisce score e componenti statistiche mediane per una strategia su un dato N_Piani.
    """
    ss = df_floor[df_floor["Strategia"] == strategy]

    # riduzioni PFA HNE/HNN (mediane separate, poi media)
    pfa_hne = robust_stats(ss[ss["Asse"]=="HNE"]["Riduzione_PFA_%"])["med"]
    pfa_hnn = robust_stats(ss[ss["Asse"]=="HNN"]["Riduzione_PFA_%"])["med"]
    pfa_hmean = np.nanmean([pfa_hne, pfa_hnn])

    # riduzioni SA HNE/HNN
    sa_hne = robust_stats(ss[ss["Asse"]=="HNE"]["Riduzione_SA_%"])["med"]
    sa_hnn = robust_stats(ss[ss["Asse"]=="HNN"]["Riduzione_SA_%"])["med"]
    sa_hmean = np.nanmean([sa_hne, sa_hnn])

    # riduzione verticale (PFA) HNZ
    pfa_hnz = robust_stats(ss[ss["Asse"]=="HNZ"]["Riduzione_PFA_%"])["med"]

    score = wPFA*pfa_hmean + wSA*sa_hmean + wV*pfa_hnz
    return {
        "score": float(score) if np.isfinite(score) else np.nan,
        "PFAred_Hmean_med": float(pfa_hmean) if np.isfinite(pfa_hmean) else np.nan,
        "SAred_Hmean_med": float(sa_hmean) if np.isfinite(sa_hmean) else np.nan,
        "PFAred_HNZ_med": float(pfa_hnz) if np.isfinite(pfa_hnz) else np.nan,
    }

# -----------------------------
# 1) PLOT "EVIDENCE": riduzioni mediane vs base fissa (per piani, per strategia)
# -----------------------------
def plot_effectiveness_over_floors(df_iso):
    """
    Figura stile paper:
    - asse x: N_Piani
    - y: riduzione mediana (PFA orizz. e SA orizz.)
    - linee per ciascuna strategia (solo quelle presenti per quel piano)
    """
    # aggrega per (N_Piani, Strategia) le mediane richieste
    rows = []
    for nP in floors:
        subP = df_iso[df_iso["N_Piani"] == nP]
        strategies = list(dict.fromkeys(subP.sort_values(["Categoria","Tis"])["Strategia"].tolist()))
        for s in strategies:
            comp = compute_strategy_score(subP, s)
            rows.append({"N_Piani": int(nP), "Strategia": s, **comp})
    df_eff = pd.DataFrame(rows)
    df_eff.to_excel(os.path.join(OUT_DIR, "EFFICACIA_ISOLAMENTO_PER_PIANO.xlsx"), index=False)

    # PFA horizontal
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title("Efficacia dell’isolamento: riduzione mediana PFA orizzontale (HNE/HNN) vs base fissa", fontsize=13, fontweight="bold")
    ax.set_xlabel("Numero di piani")
    ax.set_ylabel("Riduzione PFA orizz. mediana [%]")
    ax.grid(True, ls=":", alpha=0.4)

    for s in sorted(df_eff["Strategia"].unique()):
        ss = df_eff[df_eff["Strategia"] == s].sort_values("N_Piani")
        ss = ss[np.isfinite(ss["PFAred_Hmean_med"])]
        if ss.empty:
            continue
        ax.plot(ss["N_Piani"], ss["PFAred_Hmean_med"], marker="o", lw=2, label=s)

    ax.set_xticks(sorted(df_eff["N_Piani"].unique()))
    ax.legend(fontsize=9, loc="lower right", framealpha=0.95)
    fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "EVIDENCE_PFA_Hmean_vs_FLOORS.png"), dpi=240)
    plt.close(fig)

    # SA horizontal
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title("Efficacia dell’isolamento: riduzione mediana SA orizzontale (HNE/HNN) vs base fissa", fontsize=13, fontweight="bold")
    ax.set_xlabel("Numero di piani")
    ax.set_ylabel("Riduzione SA orizz. mediana [%]")
    ax.grid(True, ls=":", alpha=0.4)

    for s in sorted(df_eff["Strategia"].unique()):
        ss = df_eff[df_eff["Strategia"] == s].sort_values("N_Piani")
        ss = ss[np.isfinite(ss["SAred_Hmean_med"])]
        if ss.empty:
            continue
        ax.plot(ss["N_Piani"], ss["SAred_Hmean_med"], marker="o", lw=2, label=s)

    ax.set_xticks(sorted(df_eff["N_Piani"].unique()))
    ax.legend(fontsize=9, loc="lower right", framealpha=0.95)
    fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "EVIDENCE_SA_Hmean_vs_FLOORS.png"), dpi=240)
    plt.close(fig)

    return df_eff

df_eff = plot_effectiveness_over_floors(df_iso)

# -----------------------------
# 2) PLOT "CHOICE MAP": per ogni N_Piani, migliore tipologia+Tis (argmax score)
# -----------------------------
def plot_choice_map(df_iso):
    """
    Mostra chiaramente, per ogni N_Piani:
    - Strategia consigliata (tipologia e Tis) in base allo score
    - barra score
    """
    best_rows = []
    for nP in floors:
        subP = df_iso[df_iso["N_Piani"] == nP]
        strategies = list(dict.fromkeys(subP.sort_values(["Categoria","Tis"])["Strategia"].tolist()))
        best = None
        for s in strategies:
            comp = compute_strategy_score(subP, s)
            if not np.isfinite(comp["score"]):
                continue
            if best is None or comp["score"] > best["score"]:
                best = {"N_Piani": int(nP), "Strategia": s, **comp}
        if best is not None:
            best_rows.append(best)

    df_best = pd.DataFrame(best_rows)
    df_best.to_excel(os.path.join(OUT_DIR, "SCELTA_CONSIGLIATA_PER_PIANO.xlsx"), index=False)

    if df_best.empty:
        return df_best

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title("Mappa di scelta: tipologia e periodo di isolamento consigliati per numero di piani", fontsize=13, fontweight="bold")
    ax.set_xlabel("Numero di piani")
    ax.set_ylabel("Score (composito, mediana)")

    x = np.arange(len(df_best))
    ax.bar(x, df_best["score"].to_numpy(dtype=float), edgecolor="black", alpha=0.85)
    ax.set_xticks(x)
    ax.set_xticklabels(df_best["N_Piani"].astype(int).astype(str).tolist())

    # annotazioni: strategia + componenti
    for i, r in df_best.iterrows():
        txt = (
            f"{r['Strategia']}\n"
            f"PFA(H)={r['PFAred_Hmean_med']:.0f}% | SA(H)={r['SAred_Hmean_med']:.0f}% | V={r['PFAred_HNZ_med']:.0f}%"
        )
        ax.text(i, r["score"], txt, ha="center", va="bottom", fontsize=9, rotation=0)

    ax.grid(True, axis="y", ls=":", alpha=0.4)
    fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "CHOICE_MAP_BEST_STRATEGY_PER_FLOOR.png"), dpi=240)
    plt.close(fig)

    return df_best

df_best = plot_choice_map(df_iso)

# -----------------------------
# 3) PLOT "DOMINANCE": isolato vs slitte per ciascun N_Piani (Pareto-like)
# -----------------------------
def plot_dominance_iso_vs_slitte(df_iso):
    """
    Per ciascun N_Piani:
    - confronta le due famiglie (ISOLATA vs SLITTE) scegliendo il migliore Tis interno a ciascuna famiglia
    - mostra un punto per famiglia nel piano (PFA_Hmean, SA_Hmean)
    """
    rows = []
    for nP in floors:
        subP = df_iso[df_iso["N_Piani"] == nP]
        for cat in ["ISOLATA", "SLITTE"]:
            subC = subP[subP["Categoria"] == cat]
            if subC.empty:
                continue
            strategies = list(dict.fromkeys(subC.sort_values("Tis")["Strategia"].tolist()))
            best = None
            for s in strategies:
                comp = compute_strategy_score(subP, s)  # score su subP ok, ma restrizione strategia
                if not np.isfinite(comp["score"]):
                    continue
                if best is None or comp["score"] > best["score"]:
                    best = {"N_Piani": int(nP), "Famiglia": category_label(cat), "Strategia": s, **comp}
            if best is not None:
                rows.append(best)

    df_dom = pd.DataFrame(rows)
    df_dom.to_excel(os.path.join(OUT_DIR, "DOMINANCE_ISOLATA_VS_SLITTE.xlsx"), index=False)

    if df_dom.empty:
        return

    for nP in floors:
        ss = df_dom[df_dom["N_Piani"] == nP]
        if ss.empty:
            continue

        fig, ax = plt.subplots(figsize=(8.5, 6.5))
        ax.set_title(f"{int(nP)} piani – confronto famiglie (best Tis per famiglia)", fontsize=13, fontweight="bold")
        ax.set_xlabel("Riduzione PFA orizz. mediana [%] (HNE/HNN)")
        ax.set_ylabel("Riduzione SA orizz. mediana [%] (HNE/HNN)")
        ax.grid(True, ls=":", alpha=0.4)

        for _, r in ss.iterrows():
            ax.scatter(r["PFAred_Hmean_med"], r["SAred_Hmean_med"], s=120, alpha=0.85)
            ax.text(r["PFAred_Hmean_med"], r["SAred_Hmean_med"],
                    f"{r['Famiglia']}\n{r['Strategia']}", fontsize=9, ha="left", va="bottom")

        # linee guida: più in alto e più a destra = meglio
        ax.axhline(0, lw=1, ls="--", alpha=0.5)
        ax.axvline(0, lw=1, ls="--", alpha=0.5)

        fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, f"{int(nP)}P_DOMINANCE_FAMIGLIE.png"), dpi=240)
        plt.close(fig)

plot_dominance_iso_vs_slitte(df_iso)

# -----------------------------
# 4) PLOT "FINAL RECOMMENDATION" (tutti i piani in una figura)
# -----------------------------
def plot_final_recommendation(df_best):
    """
    Figura finale: per N_Piani mostra la scelta raccomandata come punto sul piano (PFA_Hmean, SA_Hmean),
    con etichetta strategia (tipologia+Tis).
    """
    if df_best is None or df_best.empty:
        return

    fig, ax = plt.subplots(figsize=(10, 7))
    ax.set_title("Raccomandazione globale: strategia ottimale (tipologia + periodo) per numero di piani", fontsize=13, fontweight="bold")
    ax.set_xlabel("Riduzione PFA orizz. mediana [%] (HNE/HNN)")
    ax.set_ylabel("Riduzione SA orizz. mediana [%] (HNE/HNN)")
    ax.grid(True, ls=":", alpha=0.4)

    # uso N_Piani come dimensione marker (senza fissare colori manualmente)
    sizes = 60 + 8*df_best["N_Piani"].to_numpy(dtype=float)
    sc = ax.scatter(df_best["PFAred_Hmean_med"], df_best["SAred_Hmean_med"], s=sizes, c=df_best["N_Piani"])

    for _, r in df_best.iterrows():
        ax.text(r["PFAred_Hmean_med"], r["SAred_Hmean_med"],
                f"{int(r['N_Piani'])}P\n{r['Strategia']}", fontsize=9, ha="left", va="bottom")

    cbar = fig.colorbar(sc, ax=ax)
    cbar.set_label("N_Piani")

    fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "FINAL_RECOMMENDATION_ALL_FLOORS.png"), dpi=260)
    plt.close(fig)

plot_final_recommendation(df_best)

# -----------------------------
# 5) Nota “bradisismo”: grafico di enfasi sulla bassa frequenza (proxy)
# -----------------------------
# Qui non abbiamo direttamente banda di frequenza o drift; usiamo un proxy:
# - se la riduzione SA (spettro 5%) è alta e stabile, la domanda dinamica viene mitigata;
# - la scelta Tis più alto tende a spostare la risposta verso periodi più lunghi
#   e ridurre accelerazioni al superstruttura.
#
# Grafico: SA-riduction vs Tis, con N_Piani separati (mediana su nodi).
def plot_reduction_vs_Tis(df_iso):
    rows = []
    for nP in floors:
        subP = df_iso[df_iso["N_Piani"] == nP]
        # strategie presenti
        strategies = list(dict.fromkeys(subP.sort_values(["Categoria","Tis"])["Strategia"].tolist()))
        for s in strategies:
            ss = subP[subP["Strategia"] == s]
            # prendo mediana SA rid (media HNE/HNN)
            sa_hne = robust_stats(ss[ss["Asse"]=="HNE"]["Riduzione_SA_%"])["med"]
            sa_hnn = robust_stats(ss[ss["Asse"]=="HNN"]["Riduzione_SA_%"])["med"]
            sa_hmean = np.nanmean([sa_hne, sa_hnn])

            # tis lo estraggo da colonna Tis
            tis_vals = ss["Tis"].dropna().unique()
            tis = float(tis_vals[0]) if len(tis_vals) else np.nan

            rows.append({"N_Piani": int(nP), "Strategia": s, "Tis": tis, "SAred_Hmean_med": sa_hmean})
    d = pd.DataFrame(rows).dropna(subset=["Tis"])

    if d.empty:
        return

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title("Riduzione SA (HNE/HNN) vs periodo di isolamento Tis – indicazione per eccitazioni low-frequency", fontsize=12, fontweight="bold")
    ax.set_xlabel("Tis [s]")
    ax.set_ylabel("Riduzione SA orizz. mediana [%]")
    ax.grid(True, ls=":", alpha=0.4)

    # per ogni N_Piani una linea (solo punti disponibili)
    for nP in sorted(d["N_Piani"].unique()):
        dd = d[d["N_Piani"] == nP].sort_values("Tis")
        ax.plot(dd["Tis"], dd["SAred_Hmean_med"], marker="o", lw=2, label=f"{int(nP)} piani")

    ax.legend(fontsize=9, loc="lower right", framealpha=0.95)
    fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "SA_REDUCTION_vs_Tis_by_FLOORS.png"), dpi=240)
    plt.close(fig)

plot_reduction_vs_Tis(df_iso)

print("OK: plot 'bradisismo/decision-support' creati in:", OUT_DIR)
print("File principali:")
print(" - EVIDENCE_PFA_Hmean_vs_FLOORS.png")
print(" - EVIDENCE_SA_Hmean_vs_FLOORS.png")
print(" - CHOICE_MAP_BEST_STRATEGY_PER_FLOOR.png")
print(" - FINAL_RECOMMENDATION_ALL_FLOORS.png")
print(" - SA_REDUCTION_vs_Tis_by_FLOORS.png")
print("Tabelle:")
print(" - EFFICACIA_ISOLAMENTO_PER_PIANO.xlsx")
print(" - SCELTA_CONSIGLIATA_PER_PIANO.xlsx")
print(" - DOMINANCE_ISOLATA_VS_SLITTE.xlsx")
```

```python
!pip install pyrotd
import os
import re
import shutil
import tempfile

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.signal import welch
from scipy.integrate import cumulative_trapezoid

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e


# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81

# Confronto "cross-floors": stessa figura contiene le curve per 3/5/10/15 piani
FLOORS_TO_COMPARE = [3, 5, 10, 15]

# File unico generato dallo script SAP (Excel master)
EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"  # <-- metti path completo se serve

# finestra di analisi comune
TMAX = 30.0

# input base
INPUT_COLB = "COLB.txt"
INPUT_DT = 0.01  # dt input COLB

# spettri (solo HNE/HNN)
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05
SA_XMAX = 3.0

# PSD
PSD_FMAX = 50.0
WELCH_NPERSEG = 2048
PSD_MODE = "DB"  # "DB" oppure "LOG" oppure "LIN"
PSD_EPS = 1e-20

# Husid
HUSID_P1 = 0.05
HUSID_P2 = 0.95

# output dirs
OUTDIR = "POSTPROCESS_OUTPUT_BASE_FISSA_CROSS_FLOORS"
DIR_COMPARE = os.path.join(OUTDIR, "01_CONFRONTI_CROSS_PIANI_TH_PSD_SA")
DIR_HUSID = os.path.join(OUTDIR, "02_CONFRONTI_CROSS_PIANI_HUSID")
DIR_TABLES = os.path.join(OUTDIR, "03_TABELLE")


# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df


def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


def truncate_by_time(t: np.ndarray, x: np.ndarray, tmax: float):
    if len(t) == 0:
        return t, x
    mask = t <= tmax
    return t[mask], x[mask]


def classify_category(tipo_base: str, modello: str):
    tb = (tipo_base or "").upper()
    mn = (modello or "").upper()
    txt = tb if tb not in ("", "NAN", "NONE") else mn
    if ("RIGIDA" in txt) or ("FISSA" in txt) or ("FIXED" in txt):
        return "RIGIDA"
    if ("SLITT" in txt) or ("FRICTION" in txt):
        return "SLITTE"
    if ("ISOL" in txt) or ("ELAST" in txt):
        return "ISOLATA"
    return "ALTRO"


def floor_style(nP: int):
    """
    Stile per distinguere i piani senza imporre colori specifici.
    Usiamo linewidth e linestyle (colori default Matplotlib).
    """
    ls_map = {3: "-", 5: "--", 10: "-.", 15: ":"}
    lw_map = {3: 2.8, 5: 2.2, 10: 2.0, 15: 2.0}
    return dict(linestyle=ls_map.get(nP, "-"), linewidth=lw_map.get(nP, 2.0))


# ============================================================
# 2) INPUT: PGA + Arias (solo input)
# ============================================================
def arias_curve(acc: np.ndarray, t: np.ndarray) -> np.ndarray:
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ


def read_colb_input_metrics(path: str, dt_input: float):
    if not os.path.exists(path):
        return {}, {}

    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    if df.shape[1] < 3:
        return {}, {}

    acc = {
        "HNE": df.iloc[:, 0].to_numpy(dtype=float) * G,
        "HNN": df.iloc[:, 1].to_numpy(dtype=float) * G,
        "HNZ": df.iloc[:, 2].to_numpy(dtype=float) * G,
    }

    pga_in, arias_in = {}, {}
    for ax, a in acc.items():
        t = np.arange(len(a)) * dt_input
        t, a = truncate_by_time(t, a, TMAX)

        pga_in[ax] = float(np.max(np.abs(a))) if len(a) else 0.0
        if len(a) >= 2:
            Ia = arias_curve(a, t)
            arias_in[ax] = float(Ia[-1])
        else:
            arias_in[ax] = 0.0

    return pga_in, arias_in


# ============================================================
# 3) HUSID
# ============================================================
def husid_metrics(Ia_curve_arr: np.ndarray, t: np.ndarray, p1=0.05, p2=0.95):
    Ia_tot = float(Ia_curve_arr[-1]) if len(Ia_curve_arr) else 0.0
    if Ia_tot <= 0:
        return dict(Ia_total=Ia_tot, t1=np.nan, t2=np.nan, duration=np.nan, husid=None)

    h = Ia_curve_arr / Ia_tot

    def _time_at(p):
        idx = np.searchsorted(h, p, side="left")
        if idx <= 0:
            return float(t[0])
        if idx >= len(t):
            return float(t[-1])
        t0, t1v = float(t[idx - 1]), float(t[idx])
        h0, h1 = float(h[idx - 1]), float(h[idx])
        if h1 == h0:
            return t1v
        return t0 + (p - h0) * (t1v - t0) / (h1 - h0)

    tp1 = _time_at(p1)
    tp2 = _time_at(p2)
    return dict(Ia_total=Ia_tot, t1=tp1, t2=tp2, duration=tp2 - tp1, husid=h)


# ============================================================
# 4) LETTURA EXCEL MASTER
# ============================================================
def load_excel_master(excel_file: str):
    if not os.path.exists(excel_file):
        raise FileNotFoundError(f"❌ Excel non trovato: {excel_file}")

    xls = pd.ExcelFile(excel_file)
    if "Riepilogo" not in xls.sheet_names or "TimeHistory" not in xls.sheet_names:
        raise ValueError("❌ Fogli richiesti mancanti (Riepilogo/TimeHistory).")

    df_r = pd.read_excel(xls, sheet_name="Riepilogo")
    df_th = pd.read_excel(xls, sheet_name="TimeHistory")

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Tipo_Base", "Nodo_Tag", "Nodo_ID"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])

    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    if "Tipo_Base" not in df_r.columns:
        df_r["Tipo_Base"] = ""

    df_r["Categoria"] = df_r.apply(
        lambda r: classify_category(r.get("Tipo_Base", ""), r.get("Modello", "")),
        axis=1
    )

    return df_r, df_th


# ============================================================
# 5) PSD / SA
# ============================================================
def compute_psd_welch(t: np.ndarray, a: np.ndarray):
    if len(t) < 4:
        return np.array([]), np.array([])
    dt = float(np.mean(np.diff(t)))
    fs = 1.0 / dt if dt > 0 else 0.0
    if fs <= 0:
        return np.array([]), np.array([])

    f, Pxx = welch(a, fs=fs, nperseg=min(WELCH_NPERSEG, len(a)))
    mask = f <= PSD_FMAX
    f = f[mask]
    Pxx = Pxx[mask]

    if PSD_MODE == "DB":
        return f, 10.0 * np.log10(Pxx + PSD_EPS)
    return f, Pxx


def compute_sa(t: np.ndarray, a: np.ndarray, cache: dict, key):
    if key in cache:
        return cache[key]
    if len(t) < 2:
        sa = np.zeros_like(PERIODS)
        cache[key] = sa
        return sa
    dt = float(np.mean(np.diff(t)))
    spec = pyrotd.calc_spec_accels(dt, a, OSC_FREQS, OSC_DAMPING)
    sa = np.asarray(spec.spec_accel, dtype=float)
    cache[key] = sa
    return sa


# ============================================================
# 6) LIMITI GLOBALI (solo base fissa)
# ============================================================
def compute_global_limits(df_r, df_th, pga_input_by_axis):
    max_acc = 0.0
    for _, p in pga_input_by_axis.items():
        max_acc = max(max_acc, float(p))

    sa_cache = {}
    max_sa = 0.0

    if PSD_MODE == "DB":
        max_psd = -np.inf
    else:
        max_psd = 0.0

    keys = df_r[["N_Piani", "Nodo_ID", "Asse", "Modello"]].drop_duplicates()

    for _, k in keys.iterrows():
        mod = k["Modello"]
        node_id = k["Nodo_ID"]
        axis = k["Asse"]

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        if len(a):
            max_acc = max(max_acc, float(np.max(np.abs(a))))

        f, Pxx = compute_psd_welch(t, a)
        if len(Pxx):
            if PSD_MODE == "DB":
                max_psd = max(max_psd, float(np.max(Pxx)))
            else:
                max_psd = max(max_psd, float(np.max(Pxx)))

        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            if len(sa):
                max_sa = max(max_sa, float(np.max(sa)))

    lims = {}
    lims["th_y"] = 1.15 * max_acc if max_acc > 0 else 1.0

    if PSD_MODE == "DB":
        if np.isfinite(max_psd):
            lims["psd_ymax"] = max_psd + 3.0
            lims["psd_ymin"] = max_psd - 80.0
        else:
            lims["psd_ymax"] = 0.0
            lims["psd_ymin"] = -80.0
    else:
        lims["psd_ymax"] = 1.15 * max_psd if max_psd > 0 else 1.0
        lims["psd_ymin"] = 0.0

    lims["sa_y"] = 1.15 * max_sa if max_sa > 0 else 1.0
    return lims


# ============================================================
# 7) PLOT CROSS-FLOORS: TRIPTICO (TH/PSD/SA)
# ============================================================
def plot_triptych_cross_floors(node_tag, node_id, axis, df_r, df_th, pga_input_by_axis, sa_cache, lims):
    """
    Una figura unica: curve per 3/5/10/15 piani sovrapposte (solo base fissa).
    Per ogni nP selezioniamo 1 modello "RIGIDA" (se ce n'è più di uno: il primo in ordine alfabetico).
    """
    fig, axs = plt.subplots(1, 3, figsize=(24, 7))
    fig.suptitle(f"BASE FISSA | CONFRONTO PIANI | {node_tag} ({node_id}) | {axis} | 0–30 s",
                 fontsize=16, fontweight="bold")

    # TH: linee ±PGA input
    pga_in = float(pga_input_by_axis.get(axis, 0.0))
    if pga_in > 0:
        axs[0].axhline(+pga_in, color="gray", ls="--", lw=1.6, alpha=0.85, label="±PGA input")
        axs[0].axhline(-pga_in, color="gray", ls="--", lw=1.6, alpha=0.85)

    # Loop piani
    for nP in FLOORS_TO_COMPARE:
        df_sel = df_r[(df_r["N_Piani"] == nP) &
                      (df_r["Nodo_ID"] == node_id) &
                      (df_r["Asse"] == axis)].copy()
        if df_sel.empty:
            continue

        # scelgo un solo modello rigido per quel gruppo
        mod = sorted(df_sel["Modello"].unique().tolist())[0]
        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        st = floor_style(nP)
        label = f"{nP}P"

        axs[0].plot(t, a, label=label, **st)

        # PSD
        f, Pxx = compute_psd_welch(t, a)
        if len(f):
            axs[1].plot(f, Pxx, label=label, **st)

        # SA (solo HNE/HNN)
        if axis != "HNZ":
            sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
            axs[2].plot(PERIODS, sa, label=label, **st)

    # formatting (limiti globali)
    axs[0].set_title("Time History + ±PGA input", fontsize=13)
    axs[0].set_xlabel("Tempo [s]")
    axs[0].set_ylabel("Acc [m/s²]")
    axs[0].set_xlim(0, TMAX)
    axs[0].set_ylim(-lims["th_y"], lims["th_y"])
    axs[0].grid(True, ls=":", alpha=0.5)
    axs[0].legend(loc="upper right", fontsize=10)

    axs[1].set_title(f"PSD (Welch) su 0–30 s ({'dB' if PSD_MODE=='DB' else PSD_MODE})", fontsize=13)
    axs[1].set_xlabel("Frequenza [Hz]")
    axs[1].set_xlim(0, PSD_FMAX)
    axs[1].set_ylim(lims["psd_ymin"], lims["psd_ymax"])
    if PSD_MODE == "LOG":
        axs[1].set_yscale("log")
        axs[1].set_ylabel("PSD [unità²/Hz] (log)")
    elif PSD_MODE == "DB":
        axs[1].set_ylabel("PSD [dB]")
    else:
        axs[1].set_ylabel("PSD [unità²/Hz]")
    axs[1].grid(True, ls=":", alpha=0.5)
    axs[1].legend(loc="upper right", fontsize=10)

    if axis != "HNZ":
        axs[2].set_title("Spettro (5%) su 0–30 s", fontsize=13)
        axs[2].set_xlabel("Periodo [s]")
        axs[2].set_ylabel("SA [m/s²]")
        axs[2].set_xlim(0, SA_XMAX)
        axs[2].set_ylim(0, lims["sa_y"])
        axs[2].grid(True, ls=":", alpha=0.5)
        axs[2].legend(loc="upper right", fontsize=10)
    else:
        axs[2].set_visible(False)

    return fig


# ============================================================
# 8) PLOT CROSS-FLOORS: HUSID
# ============================================================
def plot_husid_cross_floors(node_tag, node_id, axis, df_r, df_th):
    fig, ax = plt.subplots(figsize=(12, 7))
    fig.suptitle(f"BASE FISSA | CONFRONTO PIANI | Husid (Arias norm.) | {node_tag} ({node_id}) | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    rows = []

    for nP in FLOORS_TO_COMPARE:
        df_sel = df_r[(df_r["N_Piani"] == nP) &
                      (df_r["Nodo_ID"] == node_id) &
                      (df_r["Asse"] == axis)].copy()
        if df_sel.empty:
            continue

        mod = sorted(df_sel["Modello"].unique().tolist())[0]
        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        Ia = arias_curve(a, t)
        m = husid_metrics(Ia, t, HUSID_P1, HUSID_P2)
        if m["husid"] is None:
            continue

        st = floor_style(nP)
        ax.plot(t, m["husid"], label=f"{nP}P | D5-95={m['duration']:.2f}s", **st)

        rows.append({
            "N_Piani": nP,
            "Nodo_ID": node_id,
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "Modello": mod,
            "Arias_Output_0_30s": m["Ia_total"],
            "t5_s": m["t1"],
            "t95_s": m["t2"],
            "Durata_5_95_s": m["duration"],
        })

    ax.set_xlim(0, TMAX)
    ax.set_ylim(0, 1.02)
    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Husid = Ia(t)/Ia(T)")
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="lower right", fontsize=9, framealpha=0.95)

    return fig, rows


# ============================================================
# 9) MAIN
# ============================================================
def main():
    # reset output
    if os.path.exists(OUTDIR):
        shutil.rmtree(OUTDIR)
    os.makedirs(DIR_COMPARE, exist_ok=True)
    os.makedirs(DIR_HUSID, exist_ok=True)
    os.makedirs(DIR_TABLES, exist_ok=True)

    # INPUT: PGA + Arias (solo input)
    PGA_input_by_axis, Arias_input_by_axis = read_colb_input_metrics(INPUT_COLB, INPUT_DT)
    if not PGA_input_by_axis:
        print("⚠️ COLB.txt non trovato o non valido. PGA input sarà 0 e linee non verranno mostrate.")

    # load excel master
    df_r, df_th = load_excel_master(EXCEL_MASTER)

    # filtra SOLO base fissa e SOLO i piani richiesti
    df_r = df_r[df_r["Categoria"] == "RIGIDA"].copy()
    df_r = df_r[df_r["N_Piani"].isin(FLOORS_TO_COMPARE)].copy()
    if df_r.empty:
        raise ValueError("❌ Nessun record RIGIDA trovato per i piani selezionati nell'Excel master.")

    # limiti globali uguali
    lims = compute_global_limits(df_r, df_th, PGA_input_by_axis)
    print(
        f"Limiti globali: TH ±{lims['th_y']:.3f} | PSD [{lims['psd_ymin']:.2f}, {lims['psd_ymax']:.2f}]"
        f" | SA [0, {lims['sa_y']:.3f}]"
    )

    sa_cache = {}
    husid_rows = []

    # ATTENZIONE: qui confrontiamo i piani => quindi raggruppiamo solo per (Nodo_Tag, Nodo_ID, Asse)
    groups = df_r[["Nodo_ID", "Nodo_Tag", "Asse"]].drop_duplicates()

    for _, g in groups.iterrows():
        node_id = g["Nodo_ID"]
        node_tag = g["Nodo_Tag"]
        axis = g["Asse"]

        # triptico cross-floors
        fig1 = plot_triptych_cross_floors(
            node_tag, node_id, axis,
            df_r, df_th,
            PGA_input_by_axis,
            sa_cache, lims
        )
        fig1.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig1.savefig(os.path.join(DIR_COMPARE, f"CROSSPIANI_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig1)

        # husid cross-floors
        fig2, rows = plot_husid_cross_floors(node_tag, node_id, axis, df_r, df_th)
        husid_rows.extend(rows)
        fig2.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig2.savefig(os.path.join(DIR_HUSID, f"CROSSPIANI_HUSID_{node_tag}_{node_id}_{axis}.png"), dpi=140)
        plt.close(fig2)

    # Tabelle Husid
    if husid_rows:
        df_husid = pd.DataFrame(husid_rows)
        df_husid.to_excel(os.path.join(DIR_TABLES, "CROSSPIANI_BASEFISSA_TAB_HUSID_ARIAS_DURATE_0_30s.xlsx"), index=False)

    # Tabella input (PGA + Arias)
    df_in = pd.DataFrame([
        {"Asse": ax,
         "PGA_Input_0_30s": float(PGA_input_by_axis.get(ax, 0.0)),
         "Arias_Input_0_30s": float(Arias_input_by_axis.get(ax, 0.0))}
        for ax in ["HNE", "HNN", "HNZ"]
    ])
    df_in.to_excel(os.path.join(DIR_TABLES, "TAB_INPUT_PGA_ARIAS_0_30s.xlsx"), index=False)

    # Zip finale
    with tempfile.TemporaryDirectory() as tmp_dir:
        if os.path.exists(OUTDIR):
            shutil.copytree(OUTDIR, os.path.join(tmp_dir, OUTDIR))
        if os.path.exists(INPUT_COLB):
            shutil.copy(INPUT_COLB, os.path.join(tmp_dir, INPUT_COLB))
        if os.path.exists(EXCEL_MASTER):
            shutil.copy(EXCEL_MASTER, os.path.join(tmp_dir, os.path.basename(EXCEL_MASTER)))

        final_zip_name = "FULL_REPORT_BASE_FISSA_CROSS_PIANI"
        shutil.make_archive(final_zip_name, "zip", tmp_dir)

    print("✅ Completato. Risultati archiviati in FULL_REPORT_BASE_FISSA_CROSS_PIANI.zip")


if __name__ == "__main__":
    main()
```

```python
import os
import re
import shutil
import tempfile

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e


# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81

# Confronto cross-floors (stessa figura: curve per 3/5/10/15 piani)
FLOORS_TO_COMPARE = [3, 5, 10, 15]

# Excel master generato dallo script SAP (unico file)
EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"  # <-- metti path completo se serve

# Finestra analisi comune: spettro su accelerazione 0–30 s
TMAX = 30.0
SA_XMAX = 3.0

# Spettri
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05

# Output dirs
OUTDIR = "POSTPROCESS_OUTPUT_BASE_FISSA_CROSS_PIANI_SA_ONLY"
DIR_SA = os.path.join(OUTDIR, "01_SPETTRI_CROSS_PIANI")
DIR_TABLES = os.path.join(OUTDIR, "02_TABELLE")


# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df


def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


def truncate_by_time(t: np.ndarray, x: np.ndarray, tmax: float):
    if len(t) == 0:
        return t, x
    mask = t <= tmax
    return t[mask], x[mask]


def classify_category(tipo_base: str, modello: str):
    tb = (tipo_base or "").upper()
    mn = (modello or "").upper()
    txt = tb if tb not in ("", "NAN", "NONE") else mn
    if ("RIGIDA" in txt) or ("FISSA" in txt) or ("FIXED" in txt):
        return "RIGIDA"
    if ("SLITT" in txt) or ("FRICTION" in txt):
        return "SLITTE"
    if ("ISOL" in txt) or ("ELAST" in txt):
        return "ISOLATA"
    return "ALTRO"


def floor_style(nP: int):
    """
    Stile per distinguere i piani senza imporre colori specifici (colori default Matplotlib).
    """
    ls_map = {3: "-", 5: "--", 10: "-.", 15: ":"}
    lw_map = {3: 2.8, 5: 2.3, 10: 2.1, 15: 2.1}
    return dict(linestyle=ls_map.get(nP, "-"), linewidth=lw_map.get(nP, 2.0))


# ============================================================
# 2) LETTURA EXCEL MASTER (Riepilogo + TimeHistory)
# ============================================================
def load_excel_master(excel_file: str):
    if not os.path.exists(excel_file):
        raise FileNotFoundError(f"❌ Excel non trovato: {excel_file}")

    xls = pd.ExcelFile(excel_file)
    if "Riepilogo" not in xls.sheet_names or "TimeHistory" not in xls.sheet_names:
        raise ValueError("❌ Fogli richiesti mancanti (Riepilogo/TimeHistory).")

    df_r = pd.read_excel(xls, sheet_name="Riepilogo")
    df_th = pd.read_excel(xls, sheet_name="TimeHistory")

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Tipo_Base", "Nodo_Tag", "Nodo_ID"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])

    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    if "Tipo_Base" not in df_r.columns:
        df_r["Tipo_Base"] = ""

    df_r["Categoria"] = df_r.apply(
        lambda r: classify_category(r.get("Tipo_Base", ""), r.get("Modello", "")),
        axis=1
    )

    return df_r, df_th


# ============================================================
# 3) SA (spettro) su 0–30 s
# ============================================================
def compute_sa(t: np.ndarray, a: np.ndarray, cache: dict, key):
    if key in cache:
        return cache[key]
    if len(t) < 2:
        sa = np.zeros_like(PERIODS)
        cache[key] = sa
        return sa
    dt = float(np.mean(np.diff(t)))
    spec = pyrotd.calc_spec_accels(dt, a, OSC_FREQS, OSC_DAMPING)
    sa = np.asarray(spec.spec_accel, dtype=float)
    cache[key] = sa
    return sa


def compute_global_sa_ymax(df_r, df_th):
    """
    y-limit globale per i grafici SA (solo HNE/HNN) su tutti i piani.
    """
    sa_cache = {}
    max_sa = 0.0

    keys = df_r[["N_Piani", "Nodo_Tag", "Nodo_ID", "Asse", "Modello"]].drop_duplicates()

    for _, k in keys.iterrows():
        axis = k["Asse"]
        if axis == "HNZ":
            continue

        mod = k["Modello"]
        node_id = k["Nodo_ID"]

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
        if len(sa):
            max_sa = max(max_sa, float(np.max(sa)))

    return 1.15 * max_sa if max_sa > 0 else 1.0


# ============================================================
# 4) PLOT: CROSS-FLOORS SOLO SPETTRI, SCENARI PER NODE_TAG
# ============================================================
def plot_sa_cross_floors_by_tag(node_tag: str, axis: str, df_r, df_th, sa_cache, sa_ymax: float):
    """
    Un plot per scenario = (Nodo_Tag, Asse).
    In quel plot: 1 curva per ciascun numero di piani (3/5/10/15), usando SOLO base fissa.
    Se per uno stesso (nP, node_tag, axis) ci sono più nodi/record, seleziona quello con SA max più alto
    (scelta conservativa e robusta rispetto a duplicati).
    """
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.suptitle(f"BASE FISSA | SPETTRI | Confronto Piani | Nodo: {node_tag} | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    for nP in FLOORS_TO_COMPARE:
        df_sel = df_r[(df_r["N_Piani"] == nP) &
                      (df_r["Nodo_Tag"] == node_tag) &
                      (df_r["Asse"] == axis)].copy()
        if df_sel.empty:
            continue

        # Se ci sono più righe (es. duplicati), scelgo un (Modello,Nodo_ID) "rappresentativo":
        # criterio: massimizza SA_max già calcolata nel riepilogo (se disponibile)
        if "Max_Acc_Spettro_m/s2" in df_sel.columns:
            df_sel["__sa"] = pd.to_numeric(df_sel["Max_Acc_Spettro_m/s2"], errors="coerce").fillna(0.0)
            best = df_sel.sort_values("__sa", ascending=False).iloc[0]
        else:
            best = df_sel.sort_values(["Modello", "Nodo_ID"]).iloc[0]

        mod = best["Modello"]
        node_id = best["Nodo_ID"]

        th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == str(node_id)) & (df_th["Asse"] == axis)]
        if th.empty:
            th = df_th[(df_th["Modello"] == mod) & (df_th["Nodo_ID"] == node_id) & (df_th["Asse"] == axis)]
        if th.empty or len(th) < 2:
            continue

        t = th["Tempo_s"].to_numpy(dtype=float)
        a = th["Acc_m/s2"].to_numpy(dtype=float)
        t, a = truncate_by_time(t, a, TMAX)

        sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
        st = floor_style(nP)
        ax.plot(PERIODS, sa, label=f"{nP}P", **st)

    ax.set_xlabel("Periodo [s]")
    ax.set_ylabel("SA [m/s²]")
    ax.set_xlim(0, SA_XMAX)
    ax.set_ylim(0, sa_ymax)
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="upper right", fontsize=10)

    return fig


# ============================================================
# 5) MAIN
# ============================================================
def main():
    # reset output
    if os.path.exists(OUTDIR):
        shutil.rmtree(OUTDIR)
    os.makedirs(DIR_SA, exist_ok=True)
    os.makedirs(DIR_TABLES, exist_ok=True)

    # load excel master
    df_r, df_th = load_excel_master(EXCEL_MASTER)

    # filtra SOLO base fissa e SOLO i piani richiesti
    df_r = df_r[df_r["Categoria"] == "RIGIDA"].copy()
    df_r = df_r[df_r["N_Piani"].isin(FLOORS_TO_COMPARE)].copy()
    if df_r.empty:
        raise ValueError("❌ Nessun record RIGIDA trovato per i piani selezionati nell'Excel master.")

    # SOLO assi per cui ha senso lo spettro (HNE/HNN); se vuoi includere HNZ, togli questo filtro
    df_r = df_r[df_r["Asse"].isin(["HNE", "HNN"])].copy()

    # y-limit globale uniforme
    sa_ymax = compute_global_sa_ymax(df_r, df_th)
    print(f"SA ymax globale: {sa_ymax:.3f} m/s²")

    # cache SA
    sa_cache = {}

    # scenari: (Nodo_Tag, Asse) — niente Nodo_ID
    scenarios = df_r[["Nodo_Tag", "Asse"]].drop_duplicates()

    for _, s in scenarios.iterrows():
        node_tag = str(s["Nodo_Tag"])
        axis = str(s["Asse"])

        fig = plot_sa_cross_floors_by_tag(node_tag, axis, df_r, df_th, sa_cache, sa_ymax)
        fig.tight_layout(rect=[0, 0.02, 1, 0.93])

        safe_tag = re.sub(r"[^A-Za-z0-9_\-]+", "_", node_tag)
        fig.savefig(os.path.join(DIR_SA, f"SA_CROSSPIANI_{safe_tag}_{axis}.png"), dpi=160)
        plt.close(fig)

    # tabellina: quali (Modello,Nodo_ID) sono stati usati per ogni scenario e piano
    used_rows = []
    for node_tag, axis in scenarios.itertuples(index=False):
        for nP in FLOORS_TO_COMPARE:
            df_sel = df_r[(df_r["N_Piani"] == nP) &
                          (df_r["Nodo_Tag"] == node_tag) &
                          (df_r["Asse"] == axis)].copy()
            if df_sel.empty:
                continue
            if "Max_Acc_Spettro_m/s2" in df_sel.columns:
                df_sel["__sa"] = pd.to_numeric(df_sel["Max_Acc_Spettro_m/s2"], errors="coerce").fillna(0.0)
                best = df_sel.sort_values("__sa", ascending=False).iloc[0]
            else:
                best = df_sel.sort_values(["Modello", "Nodo_ID"]).iloc[0]
            used_rows.append({
                "Nodo_Tag": node_tag,
                "Asse": axis,
                "N_Piani": int(nP),
                "Modello_usato": best["Modello"],
                "Nodo_ID_usato": best["Nodo_ID"],
                "SAmax_riepilogo": float(best.get("Max_Acc_Spettro_m/s2", np.nan))
            })
    if used_rows:
        pd.DataFrame(used_rows).to_excel(os.path.join(DIR_TABLES, "SCENARI_SCELTE_MODELLI_E_NODI.xlsx"), index=False)

    # zip finale
    with tempfile.TemporaryDirectory() as tmp_dir:
        if os.path.exists(OUTDIR):
            shutil.copytree(OUTDIR, os.path.join(tmp_dir, OUTDIR))
        if os.path.exists(EXCEL_MASTER):
            shutil.copy(EXCEL_MASTER, os.path.join(tmp_dir, os.path.basename(EXCEL_MASTER)))
        final_zip_name = "FULL_REPORT_BASE_FISSA_CROSS_PIANI_SA_ONLY"
        shutil.make_archive(final_zip_name, "zip", tmp_dir)

    print("✅ Completato. Risultati archiviati in FULL_REPORT_BASE_FISSA_CROSS_PIANI_SA_ONLY.zip")


if __name__ == "__main__":
    main()
```

```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================
INPUT_TABLE = "TAB_OUTPUT.txt"   # <- salva qui dentro la tabella incollata
OUT_PNG = "PLOT_SINTESI_RIDUZIONI.png"

FLOORS_ORDER = [3, 5, 10, 15]
AXES_ORDER = ["HNE", "HNN", "HNZ"]

# =========================
# UTILS
# =========================
def to_float(x):
    """Converte numeri con virgola decimale e notazione E."""
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = s.replace(".", "") if re.match(r"^\d{1,3}(\.\d{3})+,\d+$", s) else s  # eventuale separatore migliaia
    s = s.replace(",", ".")
    try:
        return float(s)
    except:
        return np.nan

def strategy_label(row):
    cat = str(row["Categoria"]).upper().strip()
    tis = row.get("Tis", 0.0)
    if cat == "RIGIDA":
        return "Base Fissa"
    if cat == "ISOLATA":
        return f"Isolato T={tis:g}s"
    if cat == "SLITTE":
        return f"Isolato+Slitte T={tis:g}s"
    return f"{cat} T={tis:g}s"

def sort_strategy_key(s):
    """Ordine: Base Fissa -> Slitte (2,2.5,3) -> Isolato (2,2.5,3) -> altri."""
    s_up = s.upper()
    if "BASE FISSA" in s_up:
        return (0, 0)
    if "SLITTE" in s_up:
        t = float(re.findall(r"([0-9]+(?:\.[0-9]+)?)", s.replace(",", "."))[0])
        return (1, t)
    if "ISOLATO T=" in s_up and "SLITTE" not in s_up:
        t = float(re.findall(r"([0-9]+(?:\.[0-9]+)?)", s.replace(",", "."))[0])
        return (2, t)
    return (9, 999)

# =========================
# LOAD + CLEAN
# =========================
# prova lettura come TSV (il tuo incolla è tab-separated)
df = pd.read_csv(INPUT_TABLE, sep=r"\t", engine="python")

# normalizza colonne (nel tuo header ci sono esattamente queste)
needed = ["N_Piani","Nodo_ID","Nodo_Tag","Asse","Modello","Categoria","Tis",
          "PFA_Output_0_30s","SA_Max_0_30s","Arias_Output_0_30s"]
missing = [c for c in needed if c not in df.columns]
if missing:
    raise ValueError(f"Colonne mancanti: {missing}")

# tipi
df["N_Piani"] = pd.to_numeric(df["N_Piani"], errors="coerce").astype("Int64")
df["Tis"] = df["Tis"].apply(to_float)

for c in ["PFA_Output_0_30s","SA_Max_0_30s","Arias_Output_0_30s"]:
    df[c] = df[c].apply(to_float)

df["Categoria"] = df["Categoria"].astype(str).str.strip().str.upper()
df["Asse"] = df["Asse"].astype(str).str.strip().str.upper()
df["Nodo_Tag"] = df["Nodo_Tag"].astype(str).str.strip()

# etichetta strategia
df["Strategia"] = df.apply(strategy_label, axis=1)

# tieni solo piani e assi attesi (se presenti)
df = df[df["N_Piani"].isin(FLOORS_ORDER) & df["Asse"].isin(AXES_ORDER)].copy()

# =========================
# RIDUZIONI vs BASE FISSA
# =========================
# Media sui nodi per (N_Piani, Asse, Strategia)
agg = (df.groupby(["N_Piani","Asse","Strategia"], as_index=False)
         .agg(PFA=("PFA_Output_0_30s","mean"),
              SA =("SA_Max_0_30s","mean"),
              ARIAS=("Arias_Output_0_30s","mean")))

# baseline "Base Fissa" per (N_Piani, Asse)
base = agg[agg["Strategia"]=="Base Fissa"][["N_Piani","Asse","PFA","SA","ARIAS"]].rename(
    columns={"PFA":"PFA_base","SA":"SA_base","ARIAS":"ARIAS_base"}
)

agg = agg.merge(base, on=["N_Piani","Asse"], how="left")

# riduzioni %
agg["Red_PFA_%"] = np.where(agg["PFA_base"]>0, (1 - agg["PFA"]/agg["PFA_base"])*100, np.nan)
# SA solo HNE/HNN: su HNZ mette NaN
agg["Red_SA_%"]  = np.where((agg["Asse"]!="HNZ") & (agg["SA_base"]>0),
                            (1 - agg["SA"]/agg["SA_base"])*100, np.nan)
agg["Red_ARIAS_%"] = np.where(agg["ARIAS_base"]>0, (1 - agg["ARIAS"]/agg["ARIAS_base"])*100, np.nan)

# =========================
# PLOT SINTESI UNICO (3 pannelli)
# =========================
# strategie (escludo base fissa, perché riduzione=0 sempre)
strategies = sorted([s for s in agg["Strategia"].unique() if s!="Base Fissa"], key=sort_strategy_key)

fig, axes = plt.subplots(3, 1, figsize=(14, 14), sharex=True)

panels = [
    ("Red_PFA_%",   "Riduzione PFA output vs Base Fissa (%)  | media sui nodi", True),
    ("Red_SA_%",    "Riduzione SAmax output vs Base Fissa (%) | HNE/HNN, media sui nodi", True),
    ("Red_ARIAS_%", "Riduzione Arias output vs Base Fissa (%) | media sui nodi", True),
]

for ax, (col, title, clamp01) in zip(axes, panels):
    for s in strategies:
        sub = agg[agg["Strategia"]==s].copy()
        # curva “sintesi”: media anche sugli assi (se vuoi mantenerli separati, vedi nota sotto)
        # Qui faccio media per (N_Piani) pesando uguale i 3 assi; SA ignora HNZ per via dei NaN
        curve = (sub.groupby("N_Piani")[col].mean()).reindex(FLOORS_ORDER)

        ax.plot(FLOORS_ORDER, curve.values, marker="o", linewidth=2, label=s)

    ax.set_title(title)
    ax.set_ylabel("Riduzione [%]")
    ax.grid(True, linestyle=":", alpha=0.6)
    ax.set_ylim(0, 100)

axes[-1].set_xlabel("Numero di piani")

# legenda unica
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=2, framealpha=0.95)

fig.tight_layout(rect=[0, 0.02, 1, 0.93])
fig.savefig(OUT_PNG, dpi=180)
plt.close(fig)

print(f"✅ Salvato: {OUT_PNG}")
```

```python
import os
import re
import shutil
import tempfile

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.integrate import cumulative_trapezoid

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e


# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81

# Confronto cross-floors (stessa figura: curve per 3/5/10/15 piani)
FLOORS_TO_COMPARE = [3, 5, 10, 15]

# Excel master generato dallo script SAP (unico file)
EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"  # <-- metti path completo se serve

# Input ground motion (per PGA + Arias di input)
INPUT_COLB = "COLB.txt"
INPUT_DT = 0.01  # dt input COLB

# Finestra analisi comune
TMAX = 30.0
SA_XMAX = 3.0

# Spettri (5% damping)
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05

# Husid (durata Arias 5–95%)
HUSID_P1 = 0.05
HUSID_P2 = 0.95

# Output dirs
OUTDIR = "POSTPROCESS_OUTPUT_BASE_FISSA_PRIMA_SEZIONE"
DIR_TH = os.path.join(OUTDIR, "01_TIME_HISTORIES_PGA")
DIR_SA = os.path.join(OUTDIR, "02_SPETTRI_DI_PIANO")
DIR_HUSID = os.path.join(OUTDIR, "03_HUSID")
DIR_ARIAS = os.path.join(OUTDIR, "04_ARIAS_INTENSITY")
DIR_IDR = os.path.join(OUTDIR, "05_IDR")
DIR_TABLES = os.path.join(OUTDIR, "06_TABELLE")


# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df


def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


def truncate_by_time(t: np.ndarray, x: np.ndarray, tmax: float):
    if len(t) == 0:
        return t, x
    mask = t <= tmax
    return t[mask], x[mask]


def classify_category(tipo_base: str, modello: str):
    tb = (tipo_base or "").upper()
    mn = (modello or "").upper()
    txt = tb if tb not in ("", "NAN", "NONE") else mn
    if ("RIGIDA" in txt) or ("FISSA" in txt) or ("FIXED" in txt):
        return "RIGIDA"
    if ("SLITT" in txt) or ("FRICTION" in txt):
        return "SLITTE"
    if ("ISOL" in txt) or ("ELAST" in txt):
        return "ISOLATA"
    return "ALTRO"


def floor_style(nP: int):
    ls_map = {3: "-", 5: "--", 10: "-.", 15: ":"}
    lw_map = {3: 2.8, 5: 2.3, 10: 2.1, 15: 2.1}
    return dict(linestyle=ls_map.get(nP, "-"), linewidth=lw_map.get(nP, 2.0))


def safe_name(s: str):
    return re.sub(r"[^A-Za-z0-9_\-]+", "_", str(s))


# ============================================================
# 2) INPUT: PGA + Arias (solo input)
# ============================================================
def arias_curve(acc: np.ndarray, t: np.ndarray) -> np.ndarray:
    """Ia(t) = pi/(2g) * ∫ a^2 dt"""
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ


def read_colb_input_metrics(path: str, dt_input: float):
    """
    COLB.txt: 3 colonne in g -> conversione a m/s².
    Output:
      - pga_in[ax] = max|a(t)| su 0–TMAX
      - Ia_curve_in[ax] = Ia(t) su 0–TMAX (array)
      - t_in = timeline comune (array)
    """
    if not os.path.exists(path):
        return {}, {}, np.array([])

    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    if df.shape[1] < 3:
        return {}, {}, np.array([])

    acc = {
        "HNE": df.iloc[:, 0].to_numpy(dtype=float) * G,
        "HNN": df.iloc[:, 1].to_numpy(dtype=float) * G,
        "HNZ": df.iloc[:, 2].to_numpy(dtype=float) * G,
    }

    # timeline (assumo stessa lunghezza per tutte le colonne)
    n = len(df)
    t = np.arange(n) * dt_input
    t, _ = truncate_by_time(t, t, TMAX)

    pga_in = {}
    Ia_in = {}
    for ax, a in acc.items():
        a = a[:len(t)]
        pga_in[ax] = float(np.max(np.abs(a))) if len(a) else 0.0
        Ia_in[ax] = arias_curve(a, t) if len(a) >= 2 else np.zeros_like(t)

    return pga_in, Ia_in, t


# ============================================================
# 3) LETTURA EXCEL MASTER (incl. IDR)
# ============================================================
def load_excel_master(excel_file: str):
    if not os.path.exists(excel_file):
        raise FileNotFoundError(f"❌ Excel non trovato: {excel_file}")

    xls = pd.ExcelFile(excel_file)

    needed = {"Riepilogo", "TimeHistory", "IDR_Riepilogo", "IDR_TimeHistory"}
    missing = [s for s in needed if s not in xls.sheet_names]
    if missing:
        raise ValueError(f"❌ Fogli mancanti nel master: {missing}")

    df_r = pd.read_excel(xls, sheet_name="Riepilogo")
    df_th = pd.read_excel(xls, sheet_name="TimeHistory")
    df_idr_sum = pd.read_excel(xls, sheet_name="IDR_Riepilogo")
    df_idr_th = pd.read_excel(xls, sheet_name="IDR_TimeHistory")

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Tipo_Base", "Nodo_Tag", "Nodo_ID"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])
    df_idr_sum = _strip_cols(df_idr_sum, ["Modello", "Tipo_Base", "Asse"])
    df_idr_th = _strip_cols(df_idr_th, ["Modello", "Tipo_Base", "Asse"])

    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    # IDR Asse: nel tuo export è U1/U2 (già), ma standardizziamo comunque se serve
    if "Asse" in df_idr_sum.columns:
        df_idr_sum["Asse"] = axis_standardize(df_idr_sum["Asse"])
    if "Asse" in df_idr_th.columns:
        df_idr_th["Asse"] = axis_standardize(df_idr_th["Asse"])

    if "Tipo_Base" not in df_r.columns:
        df_r["Tipo_Base"] = ""

    df_r["Categoria"] = df_r.apply(
        lambda r: classify_category(r.get("Tipo_Base", ""), r.get("Modello", "")),
        axis=1
    )

    # Tipicamente TimeHistory non ha N_Piani: li recupero da df_r tramite join su Modello
    if "N_Piani" not in df_th.columns and "N_Piani" in df_r.columns:
        map_np = df_r[["Modello", "N_Piani"]].drop_duplicates()
        df_th = df_th.merge(map_np, on="Modello", how="left")

    return df_r, df_th, df_idr_sum, df_idr_th


# ============================================================
# 4) SA (spettro) + selezione scenario (Nodo_Tag, Asse, N_Piani)
# ============================================================
def compute_sa(t: np.ndarray, a: np.ndarray, cache: dict, key):
    if key in cache:
        return cache[key]
    if len(t) < 2:
        sa = np.zeros_like(PERIODS)
        cache[key] = sa
        return sa
    dt = float(np.mean(np.diff(t)))
    spec = pyrotd.calc_spec_accels(dt, a, OSC_FREQS, OSC_DAMPING)
    sa = np.asarray(spec.spec_accel, dtype=float)
    cache[key] = sa
    return sa


def select_best_record(df_r, nP: int, node_tag: str, axis: str):
    """
    Selezione point-based fisica:
    - stesso Nodo_Tag
    - stessa direzione
    - stesso piano
    - NESSUNA envelope
    - NESSUN worst-case
    - scelta neutra e deterministica in caso di duplicati
    """
    sub = df_r[(df_r["N_Piani"] == nP) &
               (df_r["Nodo_Tag"] == node_tag) &
               (df_r["Asse"] == axis)].copy()

    if sub.empty:
        return None

    # normalizzazione Nodo_ID
    sub["Nodo_ID"] = sub["Nodo_ID"].astype(str).str.strip()
    sub["__nid_num"] = pd.to_numeric(sub["Nodo_ID"], errors="coerce")

    # scelta stabile e neutra (non fisicamente distorsiva)
    sub = sub.sort_values(["Nodo_ID", "__nid_num", "Modello"]).drop(columns="__nid_num")

    return sub.iloc[0]


def get_timehistory(df_th, modello: str, node_id, axis: str):
    th = df_th[(df_th["Modello"] == modello) &
               (df_th["Asse"] == axis) &
               (df_th["Nodo_ID"] == str(node_id))]
    if th.empty:
        th = df_th[(df_th["Modello"] == modello) &
                   (df_th["Asse"] == axis) &
                   (df_th["Nodo_ID"] == node_id)]
    if th.empty or len(th) < 2:
        return None, None
    t = th["Tempo_s"].to_numpy(dtype=float)
    a = th["Acc_m/s2"].to_numpy(dtype=float)
    t, a = truncate_by_time(t, a, TMAX)
    return t, a


# ============================================================
# 5) HUSID + ARIAS (output)
# ============================================================
def husid_metrics(Ia_curve_arr: np.ndarray, t: np.ndarray, p1=0.05, p2=0.95):
    Ia_tot = float(Ia_curve_arr[-1]) if len(Ia_curve_arr) else 0.0
    if Ia_tot <= 0:
        return dict(Ia_total=Ia_tot, t1=np.nan, t2=np.nan, duration=np.nan, husid=None)

    h = Ia_curve_arr / Ia_tot

    def _time_at(p):
        idx = np.searchsorted(h, p, side="left")
        if idx <= 0:
            return float(t[0])
        if idx >= len(t):
            return float(t[-1])
        t0, t1v = float(t[idx - 1]), float(t[idx])
        h0, h1 = float(h[idx - 1]), float(h[idx])
        if h1 == h0:
            return t1v
        return t0 + (p - h0) * (t1v - t0) / (h1 - h0)

    tp1 = _time_at(p1)
    tp2 = _time_at(p2)
    return dict(Ia_total=Ia_tot, t1=tp1, t2=tp2, duration=tp2 - tp1, husid=h)


# ============================================================
# 6) PLOT: Time histories + PGA input
# ============================================================
def plot_acc_cross_floors(node_tag: str, axis: str, df_r, df_th, pga_input_by_axis):
    fig, ax = plt.subplots(figsize=(12, 6))
    fig.suptitle(f"BASE FISSA | Accelerazioni (0–30 s) | Nodo: {node_tag} | {axis}",
                 fontsize=14, fontweight="bold")

    # linee PGA input
    pga_in = float(pga_input_by_axis.get(axis, 0.0))
    if pga_in > 0:
        ax.axhline(+pga_in, ls="--", lw=1.8, alpha=0.8, color="gray", label="±PGA input")
        ax.axhline(-pga_in, ls="--", lw=1.8, alpha=0.8, color="gray")

    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None:
            continue
        mod = best["Modello"]
        node_id = best["Nodo_ID"]

        t, a = get_timehistory(df_th, mod, node_id, axis)
        if t is None:
            continue

        st = floor_style(int(nP))
        ax.plot(t, a, label=f"{nP}P", **st)

    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Acc [m/s²]")
    ax.set_xlim(0, TMAX)
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="upper right", fontsize=10)
    return fig


# ============================================================
# 7) PLOT: Spettri di piano (SA) cross-floors
# ============================================================
def compute_global_sa_ymax(df_r, df_th):
    sa_cache = {}
    max_sa = 0.0
    keys = df_r[["N_Piani", "Nodo_Tag", "Nodo_ID", "Asse", "Modello"]].drop_duplicates()

    for _, k in keys.iterrows():
        axis = k["Asse"]
        if axis == "HNZ":
            continue
        mod = k["Modello"]
        node_id = k["Nodo_ID"]

        t, a = get_timehistory(df_th, mod, node_id, axis)
        if t is None or len(t) < 2:
            continue

        sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
        if len(sa):
            max_sa = max(max_sa, float(np.max(sa)))

    return 1.15 * max_sa if max_sa > 0 else 1.0


def plot_sa_cross_floors(node_tag: str, axis: str, df_r, df_th, sa_cache, sa_ymax: float):
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.suptitle(f"BASE FISSA | Spettri di Piano (0–30 s) | Nodo: {node_tag} | {axis}",
                 fontsize=14, fontweight="bold")

    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None:
            continue
        mod = best["Modello"]
        node_id = best["Nodo_ID"]

        t, a = get_timehistory(df_th, mod, node_id, axis)
        if t is None:
            continue

        sa = compute_sa(t, a, sa_cache, (mod, str(node_id), axis))
        st = floor_style(int(nP))
        ax.plot(PERIODS, sa, label=f"{nP}P", **st)

    ax.set_xlabel("Periodo [s]")
    ax.set_ylabel("SA [m/s²]")
    ax.set_xlim(0, SA_XMAX)
    ax.set_ylim(0, sa_ymax)
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="upper right", fontsize=10)
    return fig


# ============================================================
# 8) PLOT: Husid + Arias(t) cross-floors
# ============================================================
def plot_husid_cross_floors(node_tag: str, axis: str, df_r, df_th):
    fig, ax = plt.subplots(figsize=(12, 6))
    fig.suptitle(f"BASE FISSA | Husid (Ia/IaTot) | Nodo: {node_tag} | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    rows = []
    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None:
            continue

        mod = best["Modello"]
        node_id = best["Nodo_ID"]

        t, a = get_timehistory(df_th, mod, node_id, axis)
        if t is None or len(t) < 2:
            continue

        Ia = arias_curve(a, t)
        m = husid_metrics(Ia, t, HUSID_P1, HUSID_P2)
        if m["husid"] is None:
            continue

        st = floor_style(int(nP))
        ax.plot(t, m["husid"], label=f"{nP}P | D5-95={m['duration']:.2f}s", **st)

        rows.append({
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "N_Piani": int(nP),
            "Modello": mod,
            "Nodo_ID": node_id,
            "Arias_Output_0_30s": float(m["Ia_total"]),
            "t5_s": float(m["t1"]),
            "t95_s": float(m["t2"]),
            "Durata_5_95_s": float(m["duration"]),
        })

    ax.set_xlim(0, TMAX)
    ax.set_ylim(0, 1.02)
    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Husid = Ia(t)/Ia(T)")
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="lower right", fontsize=9, framealpha=0.95)
    return fig, rows


def plot_arias_time_cross_floors(node_tag: str, axis: str, df_r, df_th, Ia_input_curve=None, t_input=None):
    fig, ax = plt.subplots(figsize=(12, 6))
    fig.suptitle(f"BASE FISSA | Intensità di Arias Ia(t) | Nodo: {node_tag} | {axis} | 0–30 s",
                 fontsize=14, fontweight="bold")

    # curva input (se disponibile)
    if Ia_input_curve is not None and t_input is not None and len(t_input) == len(Ia_input_curve):
        ax.plot(t_input, Ia_input_curve, lw=2.2, ls="--", alpha=0.85, label="Input Ia(t)")

    rows = []
    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None:
            continue

        mod = best["Modello"]
        node_id = best["Nodo_ID"]

        t, a = get_timehistory(df_th, mod, node_id, axis)
        if t is None or len(t) < 2:
            continue

        Ia = arias_curve(a, t)
        st = floor_style(int(nP))
        ax.plot(t, Ia, label=f"{nP}P", **st)

        rows.append({
            "Nodo_Tag": node_tag,
            "Asse": axis,
            "N_Piani": int(nP),
            "Modello": mod,
            "Nodo_ID": node_id,
            "Arias_Output_0_30s": float(Ia[-1]) if len(Ia) else 0.0
        })

    ax.set_xlim(0, TMAX)
    ax.set_xlabel("Tempo [s]")
    ax.set_ylabel("Ia(t) [-]")
    ax.grid(True, ls=":", alpha=0.6)
    ax.legend(loc="upper left", fontsize=10)
    return fig, rows


# ============================================================
# 9) PLOT: IDR profilo + IDR(t) interpiano critico
# ============================================================
def plot_idr_for_floor(nP: int, axis: str, df_idr_sum, df_idr_th):
    """
    Per un dato nP e asse:
      - profilo IDR max lungo l’altezza (Story_i vs IDR_Max)
      - storia IDR(t) dell’interpiano critico (massimo IDR_Max)
    """
    sub = df_idr_sum[(df_idr_sum["N_Piani"] == nP) & (df_idr_sum["Asse"] == axis)].copy()
    if sub.empty:
        return None, None, None

    # profilo
    sub = sub.sort_values("Story_i")
    fig_prof, axp = plt.subplots(figsize=(7, 7))
    fig_prof.suptitle(f"BASE FISSA | Profilo IDR max | {nP} Piani | {axis}",
                      fontsize=14, fontweight="bold")
    axp.plot(sub["IDR_Max"].to_numpy(dtype=float), sub["Story_i"].to_numpy(dtype=int), marker="o")
    axp.set_xlabel("IDR max [-]")
    axp.set_ylabel("Interpiano i (1=tra base e 1° piano)")
    axp.grid(True, ls=":", alpha=0.6)

    # interpiano critico
    crit_row = sub.loc[sub["IDR_Max"].astype(float).idxmax()]
    crit_story = int(crit_row["Story_i"])

    th = df_idr_th[(df_idr_th["N_Piani"] == nP) &
                   (df_idr_th["Asse"] == axis) &
                   (df_idr_th["Story_i"] == crit_story)].copy()
    if th.empty or len(th) < 2:
        return fig_prof, None, {"Crit_Story": crit_story}

    t = th["Tempo_s"].to_numpy(dtype=float)
    idr = th["IDR"].to_numpy(dtype=float)
    t, idr = truncate_by_time(t, idr, TMAX)

    fig_th, axt = plt.subplots(figsize=(12, 5))
    fig_th.suptitle(f"BASE FISSA | IDR(t) interpiano critico i={crit_story} | {nP} Piani | {axis}",
                    fontsize=14, fontweight="bold")
    axt.plot(t, idr, lw=1.2)
    axt.set_xlabel("Tempo [s]")
    axt.set_ylabel("IDR [-]")
    axt.set_xlim(0, TMAX)
    axt.grid(True, ls=":", alpha=0.6)

    info = {
        "N_Piani": int(nP),
        "Asse": axis,
        "Crit_Story": crit_story,
        "IDR_Max_Crit": float(crit_row["IDR_Max"]),
        "IDR_95perc_Crit": float(crit_row.get("IDR_95perc", np.nan)),
        "Nodo_Basso": crit_row.get("Nodo_Basso", np.nan),
        "Nodo_Alto": crit_row.get("Nodo_Alto", np.nan),
    }
    return fig_prof, fig_th, info


# ============================================================
# 10) MAIN
# ============================================================
def main():
    # reset output
    if os.path.exists(OUTDIR):
        shutil.rmtree(OUTDIR)
    os.makedirs(DIR_TH, exist_ok=True)
    os.makedirs(DIR_SA, exist_ok=True)
    os.makedirs(DIR_HUSID, exist_ok=True)
    os.makedirs(DIR_ARIAS, exist_ok=True)
    os.makedirs(DIR_IDR, exist_ok=True)
    os.makedirs(DIR_TABLES, exist_ok=True)

    # input PGA + Arias input
    PGA_input_by_axis, Ia_input_by_axis, t_input = read_colb_input_metrics(INPUT_COLB, INPUT_DT)
    if not PGA_input_by_axis:
        print("⚠️ COLB.txt non trovato o non valido: PGA input e Ia input non verranno plottati.")

    # load excel master
    df_r, df_th, df_idr_sum, df_idr_th = load_excel_master(EXCEL_MASTER)

    # filtra SOLO base fissa e SOLO i piani richiesti
    df_r = df_r[df_r["Categoria"] == "RIGIDA"].copy()
    df_r = df_r[df_r["N_Piani"].isin(FLOORS_TO_COMPARE)].copy()
    if df_r.empty:
        raise ValueError("❌ Nessun record RIGIDA trovato per i piani selezionati nell'Excel master.")

    # IDR: (solo base fissa) e piani richiesti
    if "Tipo_Base" in df_idr_sum.columns:
        df_idr_sum = df_idr_sum[df_idr_sum["Tipo_Base"].astype(str).str.upper().str.contains("RIG", na=False)].copy()
    df_idr_sum = df_idr_sum[df_idr_sum["N_Piani"].isin(FLOORS_TO_COMPARE)].copy()

    if "Tipo_Base" in df_idr_th.columns:
        df_idr_th = df_idr_th[df_idr_th["Tipo_Base"].astype(str).str.upper().str.contains("RIG", na=False)].copy()
    df_idr_th = df_idr_th[df_idr_th["N_Piani"].isin(FLOORS_TO_COMPARE)].copy()

    # scenari per (Nodo_Tag, Asse) – per TH, Husid, Arias includo anche HNZ; per SA solo HNE/HNN
    scenarios_all = df_r[["Nodo_Tag", "Asse"]].drop_duplicates()

    # ------------- TIME HISTORIES + PGA input -------------
    for node_tag, axis in scenarios_all.itertuples(index=False):
        fig = plot_acc_cross_floors(str(node_tag), str(axis), df_r, df_th, PGA_input_by_axis)
        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_TH, f"TH_CROSSPIANI_{safe_name(node_tag)}_{axis}.png"), dpi=160)
        plt.close(fig)

    # ------------- SPETTRI DI PIANO (HNE/HNN) -------------
    df_r_sa = df_r[df_r["Asse"].isin(["HNE", "HNN"])].copy()
    sa_ymax = compute_global_sa_ymax(df_r_sa, df_th)
    print(f"SA ymax globale (HNE/HNN): {sa_ymax:.3f} m/s²")

    sa_cache = {}
    scenarios_sa = df_r_sa[["Nodo_Tag", "Asse"]].drop_duplicates()
    for node_tag, axis in scenarios_sa.itertuples(index=False):
        fig = plot_sa_cross_floors(str(node_tag), str(axis), df_r_sa, df_th, sa_cache, sa_ymax)
        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_SA, f"SA_CROSSPIANI_{safe_name(node_tag)}_{axis}.png"), dpi=160)
        plt.close(fig)

    # ------------- HUSID (HNE/HNN/HNZ) + tabella -------------
    husid_rows_all = []
    for node_tag, axis in scenarios_all.itertuples(index=False):
        fig, rows = plot_husid_cross_floors(str(node_tag), str(axis), df_r, df_th)
        husid_rows_all.extend(rows)
        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_HUSID, f"HUSID_CROSSPIANI_{safe_name(node_tag)}_{axis}.png"), dpi=160)
        plt.close(fig)
    if husid_rows_all:
        pd.DataFrame(husid_rows_all).to_excel(os.path.join(DIR_TABLES, "TAB_HUSID_DURATE_0_30s.xlsx"), index=False)

    # ------------- ARIAS Ia(t) (HNE/HNN/HNZ) + tabella -------------
    arias_rows_all = []
    for node_tag, axis in scenarios_all.itertuples(index=False):
        Ia_in = Ia_input_by_axis.get(str(axis), None) if isinstance(Ia_input_by_axis, dict) else None
        fig, rows = plot_arias_time_cross_floors(
            str(node_tag), str(axis), df_r, df_th,
            Ia_input_curve=Ia_in, t_input=t_input
        )
        arias_rows_all.extend(rows)
        fig.tight_layout(rect=[0, 0.02, 1, 0.93])
        fig.savefig(os.path.join(DIR_ARIAS, f"ARIAS_TIME_CROSSPIANI_{safe_name(node_tag)}_{axis}.png"), dpi=160)
        plt.close(fig)
    if arias_rows_all:
        pd.DataFrame(arias_rows_all).to_excel(os.path.join(DIR_TABLES, "TAB_ARIAS_OUTPUT_0_30s.xlsx"), index=False)

    # ------------- IDR (profili + IDR(t) interpiano critico) -------------
    idr_info_rows = []
    for nP in FLOORS_TO_COMPARE:
        for axis in ["HNE", "HNN"]:
            fig_prof, fig_th, info = plot_idr_for_floor(int(nP), axis, df_idr_sum, df_idr_th)
            if fig_prof is not None:
                fig_prof.tight_layout(rect=[0, 0.02, 1, 0.93])
                fig_prof.savefig(os.path.join(DIR_IDR, f"IDR_PROFILE_{nP}P_{axis}.png"), dpi=160)
                plt.close(fig_prof)
            if fig_th is not None:
                fig_th.tight_layout(rect=[0, 0.02, 1, 0.93])
                fig_th.savefig(os.path.join(DIR_IDR, f"IDR_CRIT_TH_{nP}P_{axis}.png"), dpi=160)
                plt.close(fig_th)
            if info is not None:
                idr_info_rows.append(info)

    if idr_info_rows:
        pd.DataFrame(idr_info_rows).to_excel(os.path.join(DIR_TABLES, "TAB_IDR_INTERPIANO_CRITICO.xlsx"), index=False)

    # ------------- Tabelle input (PGA + Arias) -------------
    if PGA_input_by_axis:
        df_in = pd.DataFrame([
            {
                "Asse": ax,
                "PGA_Input_0_30s": float(PGA_input_by_axis.get(ax, 0.0)),
                "Arias_Input_0_30s": float(Ia_input_by_axis.get(ax, np.array([0.0]))[-1]) if isinstance(Ia_input_by_axis.get(ax, None), np.ndarray) else 0.0
            }
            for ax in ["HNE", "HNN", "HNZ"]
        ])
        df_in.to_excel(os.path.join(DIR_TABLES, "TAB_INPUT_PGA_ARIAS_0_30s.xlsx"), index=False)

    # ------------- ZIP finale -------------
    with tempfile.TemporaryDirectory() as tmp_dir:
        if os.path.exists(OUTDIR):
            shutil.copytree(OUTDIR, os.path.join(tmp_dir, OUTDIR))
        if os.path.exists(EXCEL_MASTER):
            shutil.copy(EXCEL_MASTER, os.path.join(tmp_dir, os.path.basename(EXCEL_MASTER)))
        if os.path.exists(INPUT_COLB):
            shutil.copy(INPUT_COLB, os.path.join(tmp_dir, os.path.basename(INPUT_COLB)))

        final_zip_name = "FULL_REPORT_BASE_FISSA_PRIMA_SEZIONE"
        shutil.make_archive(final_zip_name, "zip", tmp_dir)

    print("✅ Completato. Risultati archiviati in FULL_REPORT_BASE_FISSA_PRIMA_SEZIONE.zip")


if __name__ == "__main__":
    main()
```

```python
##TABELLE RELAZIONE

import os
import numpy as np
import pandas as pd

# ============================================================
# CONFIG
# ============================================================

EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"   # path completo se serve
OUT_EXCEL = "TABELLE_RISULTATI_RELAZIONE.xlsx"

# soglie prestazionali di riferimento (per tabella IDR)
IDR_SLO_FRAGILI = 0.0033
IDR_SLO_DEFORMABILI = 0.0050

# ============================================================
# UTILS
# ============================================================

def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)


# ============================================================
# LOAD MASTER
# ============================================================

def load_master(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"File non trovato: {path}")

    xls = pd.ExcelFile(path)

    df_r = pd.read_excel(xls, sheet_name="Riepilogo")
    df_th = pd.read_excel(xls, sheet_name="TimeHistory")

    # standardizzazioni
    if "Asse" in df_r.columns:
        df_r["Asse"] = axis_standardize(df_r["Asse"])
    if "Asse" in df_th.columns:
        df_th["Asse"] = axis_standardize(df_th["Asse"])

    return df_r, df_th


# ============================================================
# 1) TABELLA IDR
# ============================================================

def build_idr_table(df_r):
    """
    Richiede che nel master siano presenti colonne:
    - N_Piani
    - Nodo_Tag
    - Nodo_ID
    - Asse
    - Max_IDR  (se presente)
    """

    if "Max_IDR" not in df_r.columns:
        print("⚠️ Max_IDR non presente nel master → tabella IDR non generata")
        return pd.DataFrame()

    tab = []

    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]

        # massimo globale
        idx = sub["Max_IDR"].idxmax()
        row = sub.loc[idx]

        tab.append({
            "N_Piani": int(nP),
            "IDR_max": float(row["Max_IDR"]),
            "Nodo_critico": row["Nodo_Tag"],
            "Nodo_ID": row["Nodo_ID"],
            "Piano_critico": row.get("Piano_IDR", np.nan),
            "SLO_fragili": IDR_SLO_FRAGILI,
            "SLO_deformabili": IDR_SLO_DEFORMABILI,
            "Margine_fragili": IDR_SLO_FRAGILI - float(row["Max_IDR"]),
            "Margine_deformabili": IDR_SLO_DEFORMABILI - float(row["Max_IDR"])
        })

    return pd.DataFrame(tab)


# ============================================================
# 2) TABELLA PFA (accelerazioni di piano)
# ============================================================

def build_pfa_table(df_r):
    tab = []

    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]

        # solo HNE/HNN per confronto
        sub = sub[sub["Asse"].isin(["HNE", "HNN"])]

        pfa_max = sub["PGA_m/s2"].max()
        row = sub.loc[sub["PGA_m/s2"].idxmax()]

        tab.append({
            "N_Piani": int(nP),
            "PFA_max_m/s2": float(pfa_max),
            "Nodo": row["Nodo_Tag"],
            "Nodo_ID": row["Nodo_ID"],
            "Asse": row["Asse"]
        })

    return pd.DataFrame(tab)


# ============================================================
# 3) TABELLA SPETTRI (SA)
# ============================================================

def build_sa_table(df_r):
    tab = []

    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]
        sub = sub[sub["Asse"].isin(["HNE", "HNN"])]

        idx = sub["Max_Acc_Spettro_m/s2"].idxmax()
        row = sub.loc[idx]

        tab.append({
            "N_Piani": int(nP),
            "SA_max_m/s2": float(row["Max_Acc_Spettro_m/s2"]),
            "Periodo_critico_s": float(row["Periodo_Max_s"]),
            "Nodo": row["Nodo_Tag"],
            "Nodo_ID": row["Nodo_ID"],
            "Asse": row["Asse"]
        })

    return pd.DataFrame(tab)


# ============================================================
# 4) TABELLA ARIAS
# ============================================================

def build_arias_table(df_r):
    if "Arias_Output_0_30s" not in df_r.columns:
        print("⚠️ Arias_Output_0_30s non presente nel master → tabella Arias non generata")
        return pd.DataFrame()

    tab = []

    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]

        idx = sub["Arias_Output_0_30s"].idxmax()
        row = sub.loc[idx]

        tab.append({
            "N_Piani": int(nP),
            "Ia_tot": float(row["Arias_Output_0_30s"]),
            "Nodo": row["Nodo_Tag"],
            "Nodo_ID": row["Nodo_ID"],
            "Asse": row["Asse"]
        })

    return pd.DataFrame(tab)


# ============================================================
# MAIN
# ============================================================

def main():
    print("📥 Caricamento master...")
    df_r, df_th = load_master(EXCEL_MASTER)

    print("📊 Generazione tabelle...")

    tab_idr = build_idr_table(df_r)
    tab_pfa = build_pfa_table(df_r)
    tab_sa = build_sa_table(df_r)
    tab_arias = build_arias_table(df_r)

    print("💾 Scrittura Excel...")

    with pd.ExcelWriter(OUT_EXCEL, engine="openpyxl") as writer:
        if not tab_idr.empty:
            tab_idr.to_excel(writer, sheet_name="TAB_IDR", index=False)

        tab_pfa.to_excel(writer, sheet_name="TAB_PFA", index=False)
        tab_sa.to_excel(writer, sheet_name="TAB_SA", index=False)

        if not tab_arias.empty:
            tab_arias.to_excel(writer, sheet_name="TAB_ARIAS", index=False)

    print("✅ Tabelle generate in:", OUT_EXCEL)


if __name__ == "__main__":
    main()
```

```python
import os
import re
import shutil
import tempfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid

try:
    import pyrotd
except ImportError as e:
    raise ImportError("pyrotd non trovato. Installare con: pip install pyrotd") from e

# ============================================================
# 0) CONFIG
# ============================================================
G = 9.81
FLOORS_TO_COMPARE = [3, 5, 10, 15]
EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"

INPUT_COLB = "COLB.txt"
INPUT_DT = 0.005  # dt input COLB (5.000E-03)

TMAX = 30.0
SA_XMAX = 3.0
PERIODS = np.linspace(0.01, 4.0, 1000)
OSC_FREQS = 1.0 / PERIODS
OSC_DAMPING = 0.05
HUSID_P1 = 0.05
HUSID_P2 = 0.95

OUTDIR = "POSTPROCESS_OUTPUT_BASE_FISSA_ALLINEATO"
DIRS = [
    "01_TIME_HISTORIES_PGA", "02_SPETTRI_DI_PIANO",
    "03_HUSID", "04_ARIAS_INTENSITY",
    "05_IDR", "06_TABELLE"
]
paths = {d.split('_')[1]: os.path.join(OUTDIR, d) for d in DIRS}

# ============================================================
# 1) UTILS
# ============================================================
def _strip_cols(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip()
    return df

def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)

def safe_name(s: str):
    return re.sub(r"[^A-Za-z0-9_\-]+", "_", str(s))

def floor_style(nP: int):
    ls_map = {3: "-", 5: "--", 10: "-.", 15: ":"}
    lw_map = {3: 2.8, 5: 2.3, 10: 2.1, 15: 2.1}
    return dict(linestyle=ls_map.get(nP, "-"), linewidth=lw_map.get(nP, 2.0))

# ============================================================
# 2) CORE: ARIAS, HUSID & ALIGNMENT
# ============================================================
def arias_curve(acc: np.ndarray, t: np.ndarray) -> np.ndarray:
    """Ia(t) = pi/(2g) * ∫ a^2 dt"""
    if len(acc) < 2: return np.zeros_like(t)
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ

def husid_curve(ia_curve: np.ndarray) -> np.ndarray:
    """Husid(t) = Ia(t) / Ia_total"""
    ia_tot = ia_curve[-1]
    if ia_tot <= 0: return np.zeros_like(ia_curve)
    return ia_curve / ia_tot

def get_timehistory(df_th, modello: str, node_id, axis: str, t_ref=None):
    """
    Estrae TH, forza la sincronizzazione allo zero (t - t[0])
    e interpola sulla timeline di riferimento (t_ref).
    """
    sub = df_th[(df_th["Modello"] == modello) &
                (df_th["Asse"] == axis) &
                (df_th["Nodo_ID"] == str(node_id))]

    if sub.empty or len(sub) < 2:
        return None, None

    t_sap = sub["Tempo_s"].to_numpy(dtype=float)
    a_sap = sub["Acc_m/s2"].to_numpy(dtype=float)

    # FORZATURA SINCRONIZZAZIONE: Azzeriamo l'origine dei tempi di SAP
    t_sap_zeroed = t_sap - t_sap[0]

    if t_ref is not None:
        # Allinea i dati SAP (spesso dt=0.001) al riferimento (dt=0.005)
        a_aligned = np.interp(t_ref, t_sap_zeroed, a_sap, left=0.0, right=0.0)
        return t_ref, a_aligned

    return t_sap_zeroed, a_sap

# ============================================================
# 3) INPUT DATA
# ============================================================
def read_colb_input_metrics(path: str, dt_input: float):
    if not os.path.exists(path):
        return {}, {}, np.array([])

    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    acc_in = {
        "HNE": df.iloc[:, 0].to_numpy() * G,
        "HNN": df.iloc[:, 1].to_numpy() * G,
        "HNZ": df.iloc[:, 2].to_numpy() * G,
    }

    n = len(df)
    t = np.arange(n) * dt_input
    mask = t <= TMAX
    t = t[mask]

    pga_in = {}
    Ia_in = {}
    for ax, a in acc_in.items():
        a = a[:len(t)]
        pga_in[ax] = np.max(np.abs(a))
        Ia_in[ax] = arias_curve(a, t)

    return pga_in, Ia_in, t

# ============================================================
# 4) PLOT FUNCTIONS
# ============================================================
def plot_husid_cross_floors(node_tag, axis, df_r, df_th, Ia_in_curve, t_ref):
    fig, ax = plt.subplots(figsize=(12, 6))
    fig.suptitle(f"HUSID PLOT (Ia/Ia_tot) | Nodo: {node_tag} | {axis}", fontsize=12, fontweight="bold")

    if Ia_in_curve is not None:
        h_in = husid_curve(Ia_in_curve)
        ax.plot(t_ref, h_in, lw=2.5, ls="--", color="black", alpha=0.5, label="Input (Base)")

    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None: continue
        t, a = get_timehistory(df_th, best["Modello"], best["Nodo_ID"], axis, t_ref=t_ref)
        if t is None: continue
        Ia = arias_curve(a, t)
        h = husid_curve(Ia)
        ax.plot(t, h, label=f"{nP}P", **floor_style(int(nP)))

    ax.set_ylabel("Husid Ratio [-]"); ax.set_xlabel("Tempo [s]")
    ax.set_ylim(0, 1.05); ax.set_xlim(0, TMAX)
    ax.grid(True, ls=":"); ax.legend(loc="lower right")
    return fig

def plot_arias_time_cross_floors(node_tag, axis, df_r, df_th, Ia_in_curve, t_ref):
    fig, ax = plt.subplots(figsize=(12, 6))
    fig.suptitle(f"ARIAS Ia(t) | Allineamento Forzato | Nodo: {node_tag} | {axis}", fontsize=12, fontweight="bold")

    if Ia_in_curve is not None:
        ax.plot(t_ref, Ia_in_curve, lw=2.5, ls="--", color="black", alpha=0.6, label="Input (Base)")

    rows = []
    for nP in FLOORS_TO_COMPARE:
        best = select_best_record(df_r, nP, node_tag, axis)
        if best is None: continue
        t, a = get_timehistory(df_th, best["Modello"], best["Nodo_ID"], axis, t_ref=t_ref)
        if t is None: continue
        Ia = arias_curve(a, t)
        ax.plot(t, Ia, label=f"{nP}P", **floor_style(int(nP)))
        rows.append({"Nodo": node_tag, "Asse": axis, "Piani": nP, "Ia_Final": Ia[-1]})

    ax.set_ylabel("Ia [m/s]"); ax.set_xlabel("Tempo [s]")
    ax.set_xlim(0, TMAX); ax.grid(True, ls=":"); ax.legend(loc="upper left")
    return fig, rows

# ============================================================
# 10) MAIN
# ============================================================
def main():
    for p in paths.values():
        if os.path.exists(p): shutil.rmtree(p, ignore_errors=True)
        os.makedirs(p, exist_ok=True)

    print(f"--- Caricamento Input COLB (dt={INPUT_DT}) ---")
    pga_in_map, ia_in_map, t_input = read_colb_input_metrics(INPUT_COLB, INPUT_DT)

    print("--- Caricamento Excel SAP ---")
    xls = pd.ExcelFile(EXCEL_MASTER)
    df_r = pd.read_excel(xls, "Riepilogo")
    df_th = pd.read_excel(xls, "TimeHistory")

    df_r = _strip_cols(df_r, ["Modello", "Asse", "Nodo_Tag", "Nodo_ID", "Tipo_Base"])
    df_th = _strip_cols(df_th, ["Modello", "Asse", "Nodo_ID"])
    df_r["Asse"] = axis_standardize(df_r["Asse"])
    df_th["Asse"] = axis_standardize(df_th["Asse"])

    df_r["Categoria"] = df_r.apply(lambda r: "RIGIDA" if "RIG" in str(r.Tipo_Base).upper() else "ALTRO", axis=1)
    df_r = df_r[(df_r["Categoria"] == "RIGIDA") & (df_r["N_Piani"].isin(FLOORS_TO_COMPARE))]

    scenarios = df_r[["Nodo_Tag", "Asse"]].drop_duplicates()
    arias_results = []

    for node_tag, axis in scenarios.itertuples(index=False):
        ia_in = ia_in_map.get(axis)

        # 1. Plot Arias
        fig_ia, rows = plot_arias_time_cross_floors(node_tag, axis, df_r, df_th, ia_in, t_input)
        fig_ia.savefig(os.path.join(paths["ARIAS"], f"ARIAS_{node_tag}_{axis}.png"), dpi=150)
        plt.close(fig_ia)
        arias_results.extend(rows)

        # 2. Plot Husid
        fig_hu = plot_husid_cross_floors(node_tag, axis, df_r, df_th, ia_in, t_input)
        fig_hu.savefig(os.path.join(paths["HUSID"], f"HUSID_{node_tag}_{axis}.png"), dpi=150)
        plt.close(fig_hu)

        # 3. Plot Time Histories
        fig_th, ax_th = plt.subplots(figsize=(12, 5))
        pga_ref = pga_in_map.get(axis, 0)
        ax_th.axhline(pga_ref, color='gray', ls='--', alpha=0.5)
        ax_th.axhline(-pga_ref, color='gray', ls='--', alpha=0.5)

        for nP in FLOORS_TO_COMPARE:
            best = select_best_record(df_r, nP, node_tag, axis)
            if best is None: continue
            t, a = get_timehistory(df_th, best["Modello"], best["Nodo_ID"], axis, t_ref=t_input)
            if t is not None:
                ax_th.plot(t, a, label=f"{nP}P", **floor_style(nP))

        ax_th.set_title(f"Accelerazioni Sincronizzate (T=0) | {node_tag} | {axis}")
        ax_th.set_ylabel("Acc [m/s²]"); ax_th.set_xlabel("Tempo [s]")
        ax_th.legend(); ax_th.grid(True, ls=":")
        fig_th.savefig(os.path.join(paths["TIME"], f"TH_{node_tag}_{axis}.png"), dpi=150)
        plt.close(fig_th)

    if arias_results:
        pd.DataFrame(arias_results).to_excel(os.path.join(paths["TABELLE"], "CONFRONTO_ARIAS.xlsx"), index=False)

    print(f"✅ Analisi completata. Risultati in: {OUTDIR}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
from scipy.integrate import cumulative_trapezoid

# ============================================================
# CONFIG
# ============================================================

EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"
INPUT_COLB = "COLB.txt" # File con le 3 componenti di input
INPUT_DT = 0.005        # Delta t del segnale di input
G = 9.81
OUT_EXCEL = "TABELLE_RISULTATI_RELAZIONE.xlsx"

# soglie prestazionali di riferimento (per tabella IDR)
IDR_SLO_FRAGILI = 0.0033
IDR_SLO_DEFORMABILI = 0.0050

# ============================================================
# UTILS & METRICS
# ============================================================

def axis_standardize(series: pd.Series) -> pd.Series:
    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    return series.replace(axis_map)

def get_arias_and_duration(acc, dt):
    """Calcola Arias totale e durata significativa 5-95%"""
    t = np.arange(len(acc)) * dt
    # Ia(t)
    integ = cumulative_trapezoid(acc**2, t, initial=0)
    ia_curve = (np.pi / (2 * G)) * integ
    ia_tot = ia_curve[-1]

    if ia_tot > 0:
        h = ia_curve / ia_tot
        t5 = t[np.searchsorted(h, 0.05)]
        t95 = t[np.searchsorted(h, 0.95)]
        duration = t95 - t5
    else:
        duration = 0

    return ia_tot, duration

# ============================================================
# LOAD INPUT BASE (COLB)
# ============================================================

def load_input_base(path, dt):
    """Legge COLB.txt e restituisce dict con Ia e Durata per asse"""
    if not os.path.exists(path):
        print(f"⚠️ Segnale di input {path} non trovato.")
        return {}

    data = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    # Colonne: 0=HNE, 1=HNN, 2=HNZ (in g)
    axes = ["HNE", "HNN", "HNZ"]
    input_metrics = {}

    for i, ax_name in enumerate(axes):
        acc = data.iloc[:, i].to_numpy() * G
        ia_tot, dur = get_arias_and_duration(acc, dt)
        input_metrics[ax_name] = {"Ia_base": ia_tot, "Durata_base": dur}

    return input_metrics

# ============================================================
# TABELLE
# ============================================================

def build_idr_table(df_r):
    if "Max_IDR" not in df_r.columns:
        return pd.DataFrame()
    tab = []
    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]
        idx = sub["Max_IDR"].idxmax()
        row = sub.loc[idx]
        tab.append({
            "N_Piani": int(nP),
            "IDR_max": float(row["Max_IDR"]),
            "Nodo_critico": row["Nodo_Tag"],
            "Piano_critico": row.get("Piano_IDR", np.nan),
            "Margine_fragili": IDR_SLO_FRAGILI - float(row["Max_IDR"]),
            "Margine_deformabili": IDR_SLO_DEFORMABILI - float(row["Max_IDR"])
        })
    return pd.DataFrame(tab)

def build_arias_table(df_r, input_base_metrics):
    """Aggiunge Ia e Durata dell'output e dell'input di base"""
    if "Arias_Output_0_30s" not in df_r.columns:
        return pd.DataFrame()

    tab = []
    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[df_r["N_Piani"] == nP]

        # Prendiamo il valore massimo di Arias per quel numero di piani
        idx = sub["Arias_Output_0_30s"].idxmax()
        row = sub.loc[idx]
        ax = row["Asse"]

        # Recupero dati base per lo stesso asse
        base = input_base_metrics.get(ax, {"Ia_base": np.nan, "Durata_base": np.nan})

        tab.append({
            "N_Piani": int(nP),
            "Asse": ax,
            "Ia_Output": float(row["Arias_Output_0_30s"]),
            "Durata_Output": float(row.get("Durata_5_95_s", np.nan)),
            "Ia_Base_Input": base["Ia_base"],
            "Durata_Base_Input": base["Durata_base"],
            "Amplificazione_Ia": float(row["Arias_Output_0_30s"]) / base["Ia_base"] if base["Ia_base"] > 0 else np.nan
        })

    return pd.DataFrame(tab)

# (Le altre funzioni build_pfa_table e build_sa_table rimangono simili)
def build_pfa_table(df_r):
    tab = []
    for nP in sorted(df_r["N_Piani"].unique()):
        sub = df_r[(df_r["N_Piani"] == nP) & (df_r["Asse"].isin(["HNE", "HNN"]))]
        if sub.empty: continue
        row = sub.loc[sub["PGA_m/s2"].idxmax()]
        tab.append({"N_Piani": int(nP), "PFA_max_m/s2": float(row["PGA_m/s2"]), "Asse": row["Asse"]})
    return pd.DataFrame(tab)

# ============================================================
# MAIN
# ============================================================

def main():
    print("📥 Caricamento dati...")
    # Caricamento Excel Master
    xls = pd.ExcelFile(EXCEL_MASTER)
    df_r = pd.read_excel(xls, sheet_name="Riepilogo")
    df_r["Asse"] = axis_standardize(df_r["Asse"])

    # Caricamento e calcolo input alla base
    input_base_metrics = load_input_base(INPUT_COLB, INPUT_DT)

    print("📊 Generazione tabelle relazione...")
    tab_idr = build_idr_table(df_r)
    tab_pfa = build_pfa_table(df_r)
    tab_arias = build_arias_table(df_r, input_base_metrics)

    print("💾 Scrittura Excel finale...")
    with pd.ExcelWriter(OUT_EXCEL, engine="openpyxl") as writer:
        if not tab_idr.empty: tab_idr.to_excel(writer, sheet_name="TAB_IDR", index=False)
        if not tab_pfa.empty: tab_pfa.to_excel(writer, sheet_name="TAB_PFA", index=False)
        if not tab_arias.empty: tab_arias.to_excel(writer, sheet_name="TAB_ARIAS", index=False)

    print(f"✅ Tabelle salvate correttamente in: {OUT_EXCEL}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid

# ============================================================
# CONFIGURAZIONE
# ============================================================
FILE_PATH = "COLB.txt"
DT = 0.005  # Delta t del file COLB (5.000E-03 s)
G = 9.81    # Accelerazione di gravità m/s2
TMAX = 30.0 # Durata massima dell'analisi

def calcola_metriche_arias(path, dt, g_const):
    # 1. Caricamento dati (3 colonne: HNE, HNN, HNZ)
    try:
        data = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    except FileNotFoundError:
        print(f"❌ Errore: File {path} non trovato.")
        return

    # Timeline
    t = np.arange(len(data)) * dt
    mask = t <= TMAX
    t = t[mask]

    asse_nomi = ["HNE (U1)", "HNN (U2)", "HNZ (U3)"]
    risultati = []

    print(f"{'Asse':<10} | {'Ia Tot [m/s]':<15} | {'Durata 5-95% [s]':<15}")
    print("-" * 45)

    for i, nome in enumerate(asse_nomi):
        # Accelerazione convertita in m/s2
        acc = data.iloc[:len(t), i].to_numpy() * g_const

        # Calcolo Intensità di Arias: Ia(t) = pi/(2g) * ∫ a^2 dt
        integ = cumulative_trapezoid(acc**2, t, initial=0.0)
        ia_curve = (np.pi / (2.0 * g_const)) * integ

        ia_tot = ia_curve[-1]

        # Calcolo Durata Significativa (Husid 5% - 95%)
        h_curve = ia_curve / ia_tot
        t5 = t[np.searchsorted(h_curve, 0.05)]
        t95 = t[np.searchsorted(h_curve, 0.95)]
        durata_5_95 = t95 - t5

        risultati.append({
            "Asse": nome,
            "Ia_tot": ia_tot,
            "Durata": durata_5_95,
            "t5": t5,
            "t95": t95,
            "ia_curve": ia_curve
        })

        print(f"{nome:<10} | {ia_tot:<15.4f} | {durata_5_95:<15.2f}")

    return t, risultati

# ============================================================
# ESECUZIONE E PLOT
# ============================================================
t, res = calcola_metriche_arias(FILE_PATH, DT, G)

# Plot di verifica per le curve di Husid (Normalizzate)

fig, ax = plt.subplots(figsize=(10, 6))
for r in res:
    # Plottiamo la curva di Husid (0 a 1)
    ax.plot(t, r['ia_curve']/r['Ia_tot'], label=f"{r['Asse']} (D={r['Durata']:.2f}s)")
    # Marker per t5 e t95
    ax.scatter([r['t5'], r['t95']], [0.05, 0.95], color='red', s=30, zorder=5)

ax.axhline(0.05, color='gray', ls='--', alpha=0.5)
ax.axhline(0.95, color='gray', ls='--', alpha=0.5)
ax.set_title("Curve di Husid - Input COLB alla Base")
ax.set_ylabel("Ia(t) / Ia_tot")
ax.set_xlabel("Tempo [s]")
ax.grid(True, ls=':')
ax.legend()
plt.show()
```

```python
import os
import re
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid
from scipy.signal import correlate

# ============================================================
# 0) CONFIGURAZIONE
# ============================================================
G = 9.81
FLOORS_TO_COMPARE = [3, 5, 10, 15]
EXCEL_MASTER = r"RISULTATI_COMPLETI.xlsx"

INPUT_COLB = "COLB.txt"
INPUT_DT = 0.005  # dt input COLB (5.000E-03)

TMAX = 30.0
OUTDIR = "POSTPROCESS_RELAZIONE_COMPLETA"
DIRS = ["01_TH", "02_HUSID", "03_ARIAS", "04_TABELLE"]
paths = {d.split('_')[1]: os.path.join(OUTDIR, d) for d in DIRS}

# ============================================================
# 1) CORE: SINCRONIZZAZIONE PERFETTA (CROSS-CORRELATION)
# ============================================================
def align_signals(ref_signal, target_signal):
    """Trova lo shift ottimale per eliminare il lag fisico/numerico"""
    ref = (ref_signal - np.mean(ref_signal)) / (np.std(ref_signal) + 1e-6)
    tgt = (target_signal - np.mean(target_signal)) / (np.std(target_signal) + 1e-6)
    corr = correlate(tgt, ref, mode='full')
    lags = np.arange(-len(ref) + 1, len(tgt))
    return lags[np.argmax(corr)]

def get_timehistory(df_th, modello, node_id, axis, t_ref, acc_ref):
    sub = df_th[(df_th["Modello"] == modello) &
                (df_th["Asse"] == axis) &
                (df_th["Nodo_ID"] == str(node_id))]
    if sub.empty or len(sub) < 2: return None, None

    t_sap = sub["Tempo_s"].to_numpy()
    a_sap = sub["Acc_m/s2"].to_numpy()

    # Sincronizzazione zero-base e interpolazione
    t_sap_zeroed = t_sap - t_sap[0]
    a_interp = np.interp(t_ref, t_sap_zeroed, a_sap, left=0.0, right=0.0)

    # Allineamento tramite correlazione per eliminare il lag residuo
    if acc_ref is not None:
        shift = align_signals(acc_ref, a_interp)
        a_interp = np.roll(a_interp, -shift)
        if shift > 0: a_interp[-shift:] = 0
        elif shift < 0: a_interp[:-shift] = 0

    return t_ref, a_interp

# ============================================================
# 2) CALCOLO ARIAS E HUSID
# ============================================================
def arias_curve(acc, t):
    integ = cumulative_trapezoid(acc**2, t, initial=0.0)
    return (np.pi / (2.0 * G)) * integ

def husid_curve(ia_curve):
    ia_tot = ia_curve[-1]
    return ia_curve / ia_tot if ia_tot > 0 else np.zeros_like(ia_curve)

def read_colb_input_metrics(path, dt_input):
    if not os.path.exists(path): return {}, {}, np.array([]), {}
    df = pd.read_csv(path, sep=r"\s+", header=None, engine="python")
    t = np.arange(len(df)) * dt_input
    mask = t <= TMAX
    t = t[mask]

    acc_map, ia_map, pga_map = {}, {}, {}
    for i, ax in enumerate(["HNE", "HNN", "HNZ"]):
        acc = df.iloc[:len(t), i].to_numpy() * G
        acc_map[ax] = acc
        ia_map[ax] = arias_curve(acc, t)
        pga_map[ax] = np.max(np.abs(acc))
    return pga_map, ia_map, t, acc_map

def floor_style(nP: int):
    colors = {3: "#1f77b4", 5: "#ff7f0e", 10: "#2ca02c", 15: "#d62728"}
    ls_map = {3: "-", 5: "--", 10: "-.", 15: ":"}
    return dict(linestyle=ls_map.get(nP, "-"), color=colors.get(nP, "black"), linewidth=2.0)

# ============================================================
# 3) MAIN
# ============================================================
def main():
    for p in paths.values():
        if os.path.exists(p): shutil.rmtree(p)
        os.makedirs(p, exist_ok=True)

    print("📥 Lettura Input COLB...")
    pga_in_map, ia_in_map, t_input, acc_in_map = read_colb_input_metrics(INPUT_COLB, INPUT_DT)

    print("📥 Lettura Excel Master...")
    xls = pd.ExcelFile(EXCEL_MASTER)
    df_r = pd.read_excel(xls, "Riepilogo")
    df_th = pd.read_excel(xls, "TimeHistory")

    axis_map = {"U1": "HNE", "U2": "HNN", "U3": "HNZ"}
    df_r["Asse"] = df_r["Asse"].astype(str).str.strip().replace(axis_map)
    df_th["Asse"] = df_th["Asse"].astype(str).str.strip().replace(axis_map)
    df_r["Nodo_Tag"] = df_r["Nodo_Tag"].astype(str).str.strip()

    df_r["Categoria"] = df_r.apply(lambda r: "RIGIDA" if "RIG" in str(r.get("Tipo_Base", "")).upper() else "ALTRO", axis=1)
    df_r = df_r[(df_r["Categoria"] == "RIGIDA") & (df_r["N_Piani"].isin(FLOORS_TO_COMPARE))]

    scenarios = df_r[["Nodo_Tag", "Asse"]].drop_duplicates()
    final_arias_results = []

    for node_tag, axis in scenarios.itertuples(index=False):
        print(f"📊 Generazione Plot: {node_tag} | {axis}")

        fig_a, ax_a = plt.subplots(figsize=(11, 6))
        fig_h, ax_h = plt.subplots(figsize=(11, 6))

        ia_in = ia_in_map.get(axis)
        acc_in = acc_in_map.get(axis)

        # Plot Input
        ax_a.plot(t_input, ia_in, color='black', lw=1.5, ls='--', label="Input (Base)")
        ax_h.plot(t_input, husid_curve(ia_in), color='black', lw=1.5, ls='--', label="Input (Base)")

        for nP in FLOORS_TO_COMPARE:
            sub = df_r[(df_r["N_Piani"] == nP) & (df_r["Nodo_Tag"] == node_tag) & (df_r["Asse"] == axis)]
            if sub.empty: continue
            best = sub.sort_values("Modello").iloc[0]

            # Sincronizzazione forzata per eliminare il lag
            t, a = get_timehistory(df_th, best["Modello"], best["Nodo_ID"], axis, t_input, acc_in)
            if t is None: continue

            ia = arias_curve(a, t)
            hu = husid_curve(ia)
            style = floor_style(nP)

            ax_a.plot(t, ia, label=f"{nP}P", **style)
            ax_h.plot(t, hu, label=f"{nP}P", **style)

            final_arias_results.append({
                "Nodo": node_tag, "Asse": axis, "Piani": nP,
                "Ia_tot": ia[-1], "Ia_input": ia_in[-1],
                "Rapporto": ia[-1]/ia_in[-1]
            })

        # Estetica e Salvataggio
        for ax, fig, name, ylabel in [(ax_a, fig_a, "ARIAS", "Ia [m/s]"), (ax_h, fig_h, "HUSID", "Ia(t)/Ia_tot")]:
            ax.set_title(f"{name} Sincronizzato | {node_tag} | {axis}", fontweight="bold")
            ax.set_xlabel("Tempo [s]"); ax.set_ylabel(ylabel)
            ax.legend(); ax.grid(True, ls=":", alpha=0.5); ax.set_xlim(0, TMAX)
            fig.savefig(os.path.join(paths[name[:5]], f"{name}_{node_tag}_{axis}.png"), dpi=160)
            plt.close(fig)

    if final_arias_results:
        pd.DataFrame(final_arias_results).to_excel(os.path.join(paths["TABELLE"], "CONFRONTO_ARIAS_INPUT_OUTPUT.xlsx"), index=False)

    print(f"✅ Analisi completata con successo in: {OUTDIR}")

if __name__ == "__main__":
    main()
```
